```python
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(12,24))
for idx, volume in enumerate(volumes):
    print ('volume ',volume)
    print ('@second',seconds[idx])  
    
    # KRUIN
    # gemeten waterdiepte @ kruin
    h_krn = pd.Series(interpolate(volume, dfH_kruin2, wave=wave,name='laagdikte', locatie='kruin'),name='gemeten waterdiepte nabij de kruin (m)')
    # gemeten stroomsnelheid bodem waterlaag @ kruin
    v_krn_bdm = pd.Series(interpolate(volume, dfV_kruinbdm2, wave=wave,name='snelheid_bodem', locatie='kruin'),name='gemeten stroomsnelheid onderkant waterlaag nabij de kruin (m/s)')
    # gemeten stroomsnelheid oppervlakte waterlaag @ kruin
    v_krn_opp = pd.Series(interpolate(volume, dfV_kruinopp2, wave=wave,name='snelheid_oppervlakte', locatie='kruin'),name='gemeten stroomsnelheid bovenkant waterlaag nabij de kruin (m/s)')    
    # gemeten stroomsnelheid @ kruin, gemiddeld boven en onderkant waterlaag
    v_krn_mean = pd.Series((v_krn_bdm + v_krn_opp)/2, name='gemeten stroomsnelheid nabij de kruin (m/s)')
    # gemeten schuifspanning @ kruin bovenkant waterlaag
    tau_krn_opp_vh = pd.Series(schuifspanning4myron(v=v_krn_opp,d=h_krn), name='gemeten schuifspanning bovenkant waterlaag nabij de kruin (pa)')
    tau_krn_opp_vh.replace([np.inf, -np.inf], np.nan, inplace=True)
    tau_krn_opp_vh.fillna(0, inplace=True)
    # gemeten schuifspanning @ kruin bovenkant waterlaag
    tau_krn_bdm_vh = pd.Series(schuifspanning4myron(v=v_krn_bdm,d=h_krn), name='gemeten schuifspanning onderkant waterlaag nabij de kruin (pa)')
    tau_krn_bdm_vh.replace([np.inf, -np.inf], np.nan, inplace=True)
    tau_krn_bdm_vh.fillna(0, inplace=True)    
#     # gemeten schuifspanning @ kruin
#     tau_krn_vh = pd.Series(schuifspanning4myron(v=v_krn_mean,d=h_krn), name='gemeten schuifspanning nabij de kruin (pa)')
#     tau_krn_vh.replace([np.inf, -np.inf], np.nan, inplace=True)
#     tau_krn_vh.fillna(0, inplace=True)
    # gemeten debiet @ kruin bovenkant waterlaag
    q_krn_opp_vh = pd.Series(v_krn_opp*h_krn, name='gemeten debiet bovenkant waterlaag nabij de kruin (m$^3$/s)')
    # gemeten debiet @ kruin bovenkant waterlaag
    q_krn_bdm_vh = pd.Series(v_krn_bdm*h_krn, name='gemeten debiet onderkant waterlaag nabij de kruin (m$^3$/s)')    
#     # gemeten debiet @ kruin
#     q_krn_vh = pd.Series(v_krn_mean*h_krn, name='gemeten debiet nabij de kruin (m$^3$/s)')
    # gemeten overtollig water volume @ kruin
    u3_krn_opp_v = pd.Series(v_krn_opp**3, index=v_krn_mean.index,name='gemeten u$^3$ bovenkant waterlaag nabij de kruin (m$^3$/s$^3$)')
    # gemeten overtollig water volume @ kruin bovenkant waterlaag
    u3_krn_bdm_v = pd.Series(v_krn_bdm**3, index=v_krn_mean.index,name='gemeten u$^3$ onderkant waterlaag nabij de kruin (m$^3$/s$^3$)')    
    # vmax op basis van bovenkant en onderkant waterlaag
    v_max = (v_krn_bdm.max() + v_krn_opp.max())/2
    # theoretische stroomsnelheid @ kruin
    u_krn,T0,Tmin = create_u(v_krn_mean,a=a,hoogteV='bodem', locatie='kruin',v_max=v_max)
    u_krn = pd.Series(u_krn,name='theoretische stroomsnelheid nabij de kruin (m/s)')
    # theoretische waterdiepte @ kruin
    d_krn = pd.Series(create_d(h_krn,T0=T0,Tmin=Tmin,b=b,locatie='kruin'),name='theoretische waterdiepte nabij de kruin (m)')
    # theoretische schuifspanning @ kruin
    tau_krn_ud = pd.Series(schuifspanning4myron(v=u_krn,d=d_krn), name='theoretische schuifspanning nabij de kruin (pa)')
    tau_krn_ud.replace([np.inf, -np.inf], np.nan, inplace=True)
    tau_krn_ud.fillna(0, inplace=True)
    # theoretisch debiet @ kruin
    q_krn_ud = pd.Series(u_krn*d_krn, name='theoretisch debiet nabij de kruin (m$^3$/s)') 
    # gemeten overtollig water volume @ kruin
    u3_krn_u = pd.Series(u_krn**3, index=u_krn.index,name='theoretische u$^3$ nabij de kruin (m$^3$/s$^3$)')
    
    # HALVERWEGE TALUD
    # gemeten waterdiepte @ halverwege talud
    h_tld = pd.Series(interpolate(volume, dfH_talud2, wave=wave,name='laagdikte', locatie='helling'),name='gemeten waterdiepte halverwege het talud (m)')
    # gemeten stroomsnelheid bodem waterlaag @ halverwege talud
    v_tld_bdm = pd.Series(interpolate(volume, dfV_taludbdm2, wave=wave,name='snelheid_bodem', locatie='helling'),name='gemeten stroomsnelheid onderkant waterlaag halverwege het talud (m/s)')
    # gemeten schuifspanning @ halverwege talud
    tau_tld_vh = pd.Series(schuifspanning4myron(v=v_tld_bdm,d=h_tld), name='gemeten schuifspanning onderkant waterlaag halverwege het talud (pa)')
    tau_tld_vh.replace([np.inf, -np.inf], np.nan, inplace=True)
    tau_tld_vh.fillna(0, inplace=True)
    # gemeten debiet @ halverwge talud
    q_tld_vh = pd.Series(v_tld_bdm*h_tld, name='gemeten debiet onderkant waterlaag halverwege het talud (m$^3$/s)')
    # gemeten overtollig water volume @ halverwege talud
    u3_tld_v = pd.Series(v_tld_bdm**3, index=v_tld_bdm.index,name='gemeten u$^3$ onderkant waterlaag halverwege het talud (m$^3$/s$^3$)')   
    
    ud_tld = s_m(v0=u_krn.max(), h0=d_krn.max(),slope=1.8,location_slope=[2.3])
    u_tldB = ud_tld.loc[(1.8,0.02,u_krn.max(),d_krn.max()),'vB'].values[0]
    d_tldB = ud_tld.loc[(1.8,0.02,u_krn.max(),d_krn.max()),'hB'].values[0]
    
    # theoretische stroomsnelheid  @ halverwege talud
    u_tld = pd.Series(create_uvB(u_tldB, u_krn.index, T0_array_talud, Tmin_array_talud, volume, a=a,locatie='helling'),name='theoretische stroomsnelheid halverwege het talud (m/s)')
    # theoretische waterdiepte @ halverwege talud
    d_tld = pd.Series(create_uhB(d_tldB, d_krn.index, T0_array_talud, Tmin_array_talud, volume, b=b,locatie='helling'),name='theoretische waterdiepte halverwege het talud (m)')
    # theoretische schuifspanning @ halverwege talud
    tau_tld_ud = pd.Series(schuifspanning4myron(v=u_tld,d=d_tld), name='theoretische schuifspanning halverwege het talud (pa)')
    tau_tld_ud.replace([np.inf, -np.inf], np.nan, inplace=True)
    tau_tld_ud.fillna(0, inplace=True)
    # theretisch debiet @ halverwege talud
    q_tld_ud = pd.Series(u_tld*d_tld, name='theoretisch debiet halverwege het talud (m$^3$/s)')
    # theoretisch overtollig water volume @ halverwege talud
    u3_tld_u = pd.Series(u_tld**3, index=u_tld.index,name='theoretische u$^3$ halverwege het talud (m$^3$/s$^3$)')    
    
    # TEEN
    ud_tnn = s_m(v0=u_krn.max(), h0=d_krn.max(),slope=1.8,location_slope=[6.45])
    u_tnnB = ud_tnn.loc[(1.8,0.02,u_krn.max(),d_krn.max()),'vB'].values[0]
    d_tnnB = ud_tnn.loc[(1.8,0.02,u_krn.max(),d_krn.max()),'hB'].values[0]    

    # theoretische stroomsnelheid @ teen
    u_tnn = pd.Series(create_uvB(u_tnnB, u_krn.index, T0_array_teen, Tmin_array_teen, volume, a=a,locatie='teen'),name='theoretische stroomsnelheid nabij de teen (m/s)')
    # theoretische waterdiepte @ halverwege talud
    d_tnn = pd.Series(create_uhB(d_tnnB, d_krn.index, T0_array_teen, Tmin_array_teen, volume, b=b,locatie='teen'),name='theoretische waterdiepte nabij de teen (m)')
    
    # theoretische schuifspanning @ teen
    tau_tnn_ud = pd.Series(schuifspanning4myron(v=u_tnn,d=d_tnn), name='theoretische schuifspanning nabij de teen (pa)')
    tau_tnn_ud.replace([np.inf, -np.inf], np.nan, inplace=True)
    tau_tnn_ud.fillna(0, inplace=True)  
    # gemeten debiet @ kruin
    q_tnn_ud = pd.Series(u_tnn*d_tnn, name='theoretisch debiet nabij de teen (m$^3$/s)')
    # theoretisch overtollig water volume @ halverwege talud
    u3_tnn_u = pd.Series(u_tnn**3, index=u_tnn.index,name='theoretische u$^3$ nabij de teen (m$^3$/s$^3$)')      
    
    df_all = pd.concat((h_krn, v_krn_opp, v_krn_bdm, tau_krn_opp_vh, tau_krn_bdm_vh, q_krn_opp_vh, q_krn_bdm_vh, u3_krn_opp_v, u3_krn_bdm_v, u_krn, d_krn, tau_krn_ud, q_krn_ud, u3_krn_u,
                        h_tld,            v_tld_bdm, tau_tld_vh,                     q_tld_vh,                   u3_tld_v,                   u_tld, d_tld, tau_tld_ud, q_tld_ud, u3_tld_u,
                                                                                                                                             u_tnn, d_tnn, tau_tnn_ud, q_tnn_ud, u3_tnn_u),axis=1)
    df_all.plot(subplots=True,ax=ax, sharex=True)
plt.tight_layout()
#plt.savefig(r'D:\Projects\Pr\3517.10\Werkfolder\2017-09-27 plots\HM_alle_vh_ud_kruin_talud_teen3.png',dpi=100)
plt.show()
```
