```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#location on drive
csv_in = r'D:\Projects\Pr\3317.20\prioritering//prioritering_ruw.csv'
csv_out = r'D:\Projects\Pr\3317.20\prioritering//prioritering_refined_v2.csv'
area = r'D:\OmsWaddenzee\trunk\fews\Config\MapLayerFiles\Baggervakken//oppervlak_baggervakken.csv'
```

```python
# read into dataframe
df_csv = pd.read_csv(csv_in, sep=';')
df_csv = df_csv[(df_csv['Datum'] == datetime.today().strftime('%Y-%m-%d'))]
df_area = pd.read_csv(area)

# join dataframes on objectcode
df_comb = pd.merge(df_csv, df_area, left_on='Baggervak', right_on='Objectcode')
df_comb['Oppervlak'] = df_comb['Oppervlak'].astype(float).apply(np.ceil)
df_comb['Oppervlakte cut tov NGD'] = df_comb['Oppervlakte cut tov NGD'].astype(float).apply(np.floor)
df_comb['Oppervlakte cut tov OHD'] = df_comb['Oppervlakte cut tov OHD'].astype(float).apply(np.floor)

# calculate the prioritering
df_comb['Percentage cut tov NGD'] = (df_comb['Oppervlakte cut tov NGD'] / df_comb['Oppervlak'] * 100).round(0)
df_comb['Percentage cut tov OHD'] = (df_comb['Oppervlakte cut tov OHD'] / df_comb['Oppervlak'] * 100).round(0)

# reorder columns 
c = df_comb.columns.tolist()
c_selection = [c[0],c[11],c[10],c[13],c[14],c[15],c[6],c[7],c[2],c[3],c[8],c[9],c[4],c[5],c[12]]

# save as csv
#df_comb[c_selection].to_csv(csv_out, sep=',', index=False)
print('end of script')


# In[ ]:
```
