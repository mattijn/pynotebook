Read yearly averages from PSMSL website. Shift the RLR data to NAP data (mm+NAP).

```python
PSMSL_ids = [32, 20, 9, 22, 23, 24, 25, 236, 1551]
#PSMSL_locations = r'D:\jupyter notebooks\2695.40\Werkzaamheden\rlr_annual.zip'
PSMSL_locations = r'P:\Pr\2695.40\Werkzaamheden\Jupyter Notebooks\Narekenen van meetreeksen\rlr_annual.zip'
#PSMSL_locations = r'C:\Users\Public\Documents\Python Scripts\rlr_annual.zip'
print(PSMSL_locations)
dataset_name = 'rlr_annual'
# these compute the rlr back to NAP (ignoring the undoing of the NAP correction)
# this means that the data are now in mm+NAP with NAP correction in 2005
main_stations = {
    20: {
        'name': 'Vlissingen', 
        'rlr2nap': lambda x: x - (6976-46)
    },
    22: {
        'name': 'Hoek van Holland', 
        'rlr2nap': lambda x:x - (6994 - 121)
    },
    23: {
        'name': 'Den Helder', 
        'rlr2nap': lambda x: x - (6988-42)
    },
    24: {
        'name': 'Delfzijl', 
        'rlr2nap': lambda x: x - (6978-155)
    },
    25: {
        'name': 'Harlingen', 
        'rlr2nap': lambda x: x - (7036-122)
    },
    32: {
        'name': 'IJmuiden', 
        'rlr2nap': lambda x: x - (7033-83)
    },
    1551: {
        'name': 'Roompot buiten',
        'rlr2nap': lambda x: x - (7011-17)
    },
    9: {
        'name': 'Maassluis',
        'rlr2nap': lambda x: x - (6983-184)
    },
    236: {
        'name': 'West-Terschelling',
        'rlr2nap': lambda x: x - (7011-54)
    },
}

# Roompot Buiten, Maassluis en West-Terschelling zijn niet-hoofdstations waarvan de data wel aan PSMSL zijn doorgegeven. 
```

Extract the data from the zipfile and process them.

```python
zf = zipfile.ZipFile(PSMSL_locations)

# this list contains a table of 
# station ID, latitude, longitude, station name, coastline code, station code, and quality flag
csvtext = zf.read('{}/filelist.txt'.format(dataset_name))

stations = pd.read_csv(
    io.BytesIO(csvtext), 
    sep=';',
    names=('id', 'lat', 'lon', 'name', 'coastline_code', 'station_code', 'quality'),
    converters={
        'name': str.strip,
        'quality': str.strip
    }
)
stations = stations.set_index('id')

# the dutch stations in the PSMSL database, make a copy
# or use stations.coastline_code == 150 for all dutch stations
selected_stations = stations.loc[PSMSL_ids].copy()
# set the main stations, this should be a list of 6 stations
selected_stations
```

```python
# each station has a number of files that you can look at.
# here we define a template for each filename

# stations that we are using for our computation
# define the name formats for the relevant files
names = {
    'datum': '{dataset}/RLR_info/{id}.txt',
    'diagram': '{dataset}/RLR_info/{id}.png',
    'url': 'http://www.psmsl.org/data/obtaining/rlr.diagrams/{id}.php',
    'data': '{dataset}/data/{id}.rlrdata',
    'doc': '{dataset}/docu/{id}.txt',
    'contact': '{dataset}/docu/{id}_auth.txt'
}
```

```python
def get_url(station, dataset):
    """return the url of the station information (diagram and datum)"""
    info = dict(
        dataset=dataset,
        id=station.name
    )
    url = names['url'].format(**info)
    return url
# fill in the dataset parameter using the global dataset_name
f = functools.partial(get_url, dataset=dataset_name)
# compute the url for each station
selected_stations['url'] = selected_stations.apply(f, axis=1)
selected_stations
```

```python
def missing2nan(value, missing=-99999):
    """convert the value to nan if the float of value equals the missing value"""
    value = float(value)
    if value == missing:
        return np.nan
    return value

def get_data(station, dataset):
    """get data for the station (pandas record) from the dataset (url)"""
    info = dict(
        dataset=dataset,
        id=station.name
    )
    bytes = zf.read(names['data'].format(**info))
    df = pd.read_csv(
        io.BytesIO(bytes), 
        sep=';', 
        names=('year', 'height', 'interpolated', 'flags'),
        converters={
            "height": lambda x: main_stations[station.name]['rlr2nap'](missing2nan(x)),
            "interpolated": str.strip,
        }
    )
    df['station'] = station.name
    return df
```

```python
# get data for all stations
f = functools.partial(get_data, dataset=dataset_name)
# look up the data for each station
selected_stations['data'] = [f(station) for _, station in selected_stations.iterrows()]
```

```python
# we now have data for each station
selected_stations[['name', 'data']]
```

```python
columns_ = [i for i in selected_stations['name']]
df2 = pd.DataFrame(index=index, columns=columns_) # PSMSL data
```

```python
for name in tqdm(columns_):
    print (name)
    df_tmp = selected_stations[selected_stations['name']==name]['data'].iloc[0][['year', 'height']]
    for row in df_tmp.iterrows():
        df2.set_value(pd.Timestamp(str(int(row[1]['year']))+'-12-31', freq='A-DEC'),name,row[1]['height'])   
df2.sort_index(inplace=True)
df2.tail() #in mm+NAP
#df2.to_csv('PSMSL_stations.csv') # RN: data wegschrijven naar csv
```

```python
# selecteer de benamingen van RWS en PSMSL
RWS_stations = ['DELFZL','DENHDR','HARLGN','HOEKVHLD','IJMDBTHVN','VLISSGN']
PSMSL_stations = ['DELFZIJL','DEN HELDER', 'HARLINGEN','HOEK VAN HOLLAND','IJMUIDEN','VLISSINGEN']

# import modules voor wegschrijven naar excel (in sheets)
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows

# zet aantal decimalen
decimals = 2

total_df = pd.DataFrame()
# open excel workbook
wb = Workbook()
for station in RWS_stations:
    wb.create_sheet(title=station)
    ws = wb.get_sheet_by_name(station)
    
    print (station)
    idx = RWS_stations.index(station)    
    RWS_sel = df[RWS_stations[idx]].copy()
    RWS_sel.name = RWS_sel.name+' (RWS, cm+NAP)'
    PSMSL_sel = (df2[PSMSL_stations[idx]]/10).copy()
    PSMSL_sel.name = PSMSL_sel.name+' (PSMSL, cm+NAP)'
    dif = RWS_sel - PSMSL_sel
    dif.name = 'verschil RWS-PSMSL (cm)'
    comb_df = pd.concat(objs=(RWS_sel,PSMSL_sel,dif), axis=1)

    comb_df.index = comb_df.index.year
    
    comb_df = comb_df.astype(float).round(decimals)    
    
    # comb_df.to_excel('stationData.xlsx', sheet_name=PSMSL_stations[idx])
    # append sheet lijkt niet te werken. vandaar dat we openpyxl direct aanroepen    
    for r in dataframe_to_rows(comb_df, index=True, header=True):
        ws.append(r)
    total_df = pd.concat(objs=(total_df, comb_df), axis=1)
    
# save total_df to excel
total_df.to_excel("stationData_singlesheet.xlsx")    
```

```python
# verwijder sheet 0 alvorens op te slaan
if wb.get_sheet_names()[0] == 'Sheet':
    wb.remove_sheet(wb.get_sheet_by_name(wb.get_sheet_names()[0]))
wb.save("stationData_multisheet.xlsx")
```

Conclusion. The yearly average values from both sources are not equal. 

Possible explanations:

1) incorrect RWS data (a comparison of the RWS data on live.waterbase.nl and the above data show that the data are correct)

2) representation of time (maybe 1 hour time difference, impact is small)

3) NAP corrections (differences after the year 2004 are equal to 'datum/NAP shift'. RWS data are not corrected, PSMSL yearly data are! use montly metric data to check this)

4) yearly averages are calculated in a different way and/or RWS has applied validations on the high-frequency data, NaN-values are treated differently (we also check this with the montly metric data)
