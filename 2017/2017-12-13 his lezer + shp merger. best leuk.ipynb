```python
import hkvsobekpy as hkv
import geopandas as gpd

his_file = r'D:\jupyter notebooks\spelen\20171212_daniel4mattijn\CALCPNT.HIS'
shp_file = r'D:\jupyter notebooks\spelen\20171212_daniel4mattijn\Sbk_Conn_n.shp'

reflow = hkv.read_his.LeesMetadata(his_file)

locaties = reflow.KrijgLokaties()
tijdstappen = reflow.KrijgTijdstappen()
parameters = reflow.KrijgParameters()
```

```python
gdf = gpd.read_file(shp_file)
gdf = gdf.rename(columns={'ID        ':'locations'})
```

```python
#uitlezen resultaten uit HIS files
parameter = []
parameter.append(parameters[2])
```

```python
df = reflow.MultiWaardenArary(
    gdf['locations'].iloc[0:100].values,
    parameters[2:4],
    startMMdd=(1,1),
    endMMdd=(12,31),
    jaarmax_as='none')
```

```python
new_column_name = 'max'
#bepaal statistiek over alle punten.
df_stat = df.max() #bepaal max van alle locs
df_stat.name = new_column_name

df_stat_reset = df_stat.reset_index()
df_pivot = df_stat_reset.pivot(index='locations', columns='parameters', values=new_column_name)
df_pivot.head()
```

```python
#merge gdf met df_stat
#pd.merge(gdf, df_pivot, how='left', right_index=True, left_on='locations')
result = gdf.merge(df_pivot, left_on='locations', right_index=True)
```
