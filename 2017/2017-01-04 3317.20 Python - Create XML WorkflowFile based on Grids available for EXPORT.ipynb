```python
# In[2]:

def prettify(elem):
    """Return a pretty-printed XML string for the Element.
    """
    rough_string = ElementTree.tostring(elem, 'utf-8')
    reparsed = minidom.parseString(rough_string)
    return reparsed.toprettyxml(indent="  ")
```

```python
asciiFiles = []
asciiFile_fullPath = []
for root, dirs, files in os.walk(rootDir):
    for file in files:
        if file.startswith('grid') and file.endswith('.asc'):
            asciiFiles.append(file)
            asciiFile_fullPath.append(os.path.join( os.path.abspath(root), file ))

asciiLocs = []
for file in asciiFiles:
    asciiLocs.append(file[0:7])

asciiLocsUnique = list(set(asciiLocs))
asciiLocsUnique.sort()


asciiFilesUnique = []
asciiFilesUnique_fullPath = []
for i in asciiLocsUnique:
    asciiFilesUnique.append(next(obj for obj in asciiFiles if obj[0:7]==i))
    asciiFilesUnique_fullPath.append(next(obj for obj in asciiFile_fullPath if obj[-26:-19]==i))
```

```python
def workflow_ex_Surveyor_geoxyz_periodiek(grids_list):
    # start xml
    top = Element('workflow')
    top.set('version','1.1')
    top.set('xmlns','http://www.wldelft.nl/fews')
    top.set('xmlns:xsi','http://www.w3.org/2001/XMLSchema-instance')
    top.set('xsi:schemaLocation','http://www.wldelft.nl/fews ../../../../Schemas/workflow.xsd')  
    
    for grid in grids_list:
        # define non grid specific variable
        var0 = SubElement(top, 'activity')
        var0sub0 = SubElement(var0, 'properties')
        var0sub0sub0 = SubElement(var0sub0, 'string')
        var0sub0sub0.set('key','location')    
        var0sub0sub0.set('value','l.'+grid)
        var0sub1 = SubElement(var0, 'runIndependent').text = 'true'
        var0sub2 = SubElement(var0, 'moduleInstanceId').text = 'ex.Surveyor.geoxyz.periodiek'
        
    return top
```

```python
# CREATE wf.pr.Surveyor.geoxyz.depthUpdate.xml
# file to save
wf_ex_Surveyor_geoxyz_periodiek = os.path.join(xmldir,'wf.ex.Surveyor.geoxyz.periodiek.xml')

top = workflow_ex_Surveyor_geoxyz_periodiek(asciiLocsUnique)

# save xml to file
with open(wf_ex_Surveyor_geoxyz_periodiek, 'w') as the_file:
    the_file.write(prettify(top))    
```
