```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib
from matplotlib import cm
import numpy as np
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
matplotlib.rcParams.update({'font.size': 22})
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%\n({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
matplotlib.rcParams.update({'font.size': 22})
df = pd.DataFrame({'Type': ['Code','Docstring','Comment','Empty'], 
                   'Number':[1709,309,183,504],
                   '%':[63.2,11.39,6.77,18.64]
                  })
ax = df.plot(kind='pie', y='Number', labels=df['Type'], figsize=(10,10),legend=False, autopct=make_autopct(df['Number']), shadow=False)
ax.set_ylabel('')
plt.tight_layout()
plt.savefig(r'D:\Projects\Pr\3629.10\werkfolder\review_week45//raw_metrics', dpi=100)
plt.show()
```
