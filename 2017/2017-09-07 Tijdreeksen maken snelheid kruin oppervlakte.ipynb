```python
def interpolate(volume, df2, wave='golf500-2000'):
    """
    Parameters
    ----------    
    volumes : int
        totaal volume (lm) waarover tijdseries bepaald moet worden
    df2 : pandas.DataFrame
        dataframe waarin de snelheid_oppervlakte per afvoergolf is bepaald vanuit de hydraulische meting    
    
    Returns
    -------
    interpolatie : pandas.Series
        geinterpoleerde tijdreeks op basis van de input volume
    """
    
    if wave == 'golf500-2000':
        # usage of golf 500 - 2000
        if volume < 1000:
            s_itp = ((volume - 500) / 500) * (df2['snelheid_oppervlakte_1000lm'] - df2['snelheid_oppervlakte_500lm']) + df2['snelheid_oppervlakte_500lm']
        elif volume > 1500:
            s_itp = ((volume - 1500) / 500) * (df2['snelheid_oppervlakte_2000lm'] - df2['snelheid_oppervlakte_1500lm']) + df2['snelheid_oppervlakte_1500lm']  
        else:
            s_itp = ((volume - 1000) / 500) * (df2['snelheid_oppervlakte_1500lm'] - df2['snelheid_oppervlakte_1000lm']) + df2['snelheid_oppervlakte_1000lm'] 
    
    if wave == 'golf500-3500':
        # usage of golf 500 - 3500
        if volume < 1000:
            s_itp = ((volume - 500) / 500) * (df2['snelheid_oppervlakte_1000lm'] - df2['snelheid_oppervlakte_500lm']) + df2['snelheid_oppervlakte_500lm']        
        elif 1000 <= volume <= 1500:
            s_itp = ((volume -1000) / 500) * (df2['snelheid_oppervlakte_1500lm'] - df2['snelheid_oppervlakte_1000lm']) + df2['snelheid_oppervlakte_1000lm']     
        elif 1500 <= volume <= 2000:
            s_itp = ((volume - 1500) / 500) * (df2['snelheid_oppervlakte_2000lm'] - df2['snelheid_oppervlakte_1500lm']) + df2['snelheid_oppervlakte_1500lm']
        elif 2000 <= volume <= 2500:
            s_itp = ((volume - 2000) / 500) * (df2['snelheid_oppervlakte_2500lm'] - df2['snelheid_oppervlakte_2000lm']) + df2['snelheid_oppervlakte_2000lm']        
        elif 2500 <= volume <= 3000:
            s_itp = ((volume - 2500) / 500) * (df2['snelheid_oppervlakte_3000lm'] - df2['snelheid_oppervlakte_2500lm']) + df2['snelheid_oppervlakte_2500lm']
        elif volume > 3000:
            s_itp = ((volume - 3000) / 500) * (df2['snelheid_oppervlakte_3500lm'] - df2['snelheid_oppervlakte_3000lm']) + df2['snelheid_oppervlakte_3000lm']
        
    interpolatie = pd.Series(s_itp,name='Geinterpoleerd (V= '+str(volume)+'l)')
    
    # zorg ervoor dat er geen negatieve waarden kunnen bestaan
    interpolatie[interpolatie < 0] = 0
    # wanneer de series de nul lijn heeft geraakt zorg ervoor dat deze op nul blijft
    if volume < 1500:
        try:
            # issue voor lagere volumes
            lzidx = np.where(np.diff(np.sign(interpolatie.values)))[0][0]
            #lzidx = interpolatie.iloc[10::].searchsorted(0, side='right')[0]-10
            #interpolatie.iloc[lzidx::] = 0
        except:
            pass
    return interpolatie
```

```python
def getTijdreeksen(hs,stepsize,dfV_kruinopp2, namef, param='snelheid_oppervlakte', loc_set='kruin'):
    volumes = hs['l/m'].values
    seconds = hs['s'].values

    lengteproef = np.arange(int(seconds[-1]/stepsize+(2000)))* stepsize
    series = pd.Series(index=lengteproef, data = 0, name=param+' (m)')
    series.index.name='time (s)'

    for idx, volume in enumerate(volumes):
        volume_interpolated = interpolate(volume, dfV_kruinopp2, wave='golf500-3500')
        tmin = series.index.searchsorted(seconds[idx], side='left')
        tmax = series.index.searchsorted(seconds[idx]+volume_interpolated.size * stepsize, side='left')
        series.iloc[tmin:tmax] = series.iloc[tmin:tmax] + volume_interpolated.values
    series.to_csv(r'D:\Projects\Pr\3517.10\Werkfolder\2017-09-07_tijdreeksen\afgeleiden/'+param+'_'+loc_set+'_'+namef+'.csv', header=True)  
    print ('ok')
    return series 
```
