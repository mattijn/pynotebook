```python
import pandas as pd
from matplotlib.dates import DayLocator, HourLocator, DateFormatter, drange
%matplotlib inline
```

```python
file_in = r'D:\Projects\RO\HydroAsAService//Tijdreeksen_Oosterscheldestations.txt'
```

```python
with open(file_in) as f:
    lines = f.readlines()[4:34]
lines = [line.rstrip('\n') for line in lines]
lines = [line.lstrip('>') for line in lines]
lines = [line.split() for line in lines]

# parm = line[0]
# dist = line[1]
# unit = line[-1]
stations = [' '.join(line[2:-1]) for line in lines]    
```

```python
stations.insert(0,'tijd')
stations.insert(0,'datum')

stations
```

```python
# df = pd.read_csv(file_in,delim_whitespace=True, skiprows=34, names = stations)
```

```python
df = pd.read_csv(file_in, delim_whitespace=True, header=None, skiprows=34)
#self.afstand      = self.implic['Cumafstand'].values  # Haal cumulatieve afstanden van locaties op
#self.locatiesrond = self.implic['Naam'].values        # Haal namen van locaties op                    

#       Voeg headers toe aan alle kolommen in het dataframe
df.columns = stations

# Converteer datum en tijd naar strings in pandas dataframe
df['datum'] = df.datum.apply (str)
df['tijd']  = df.tijd.apply (str)
df['tijd']  = df['tijd'].str.zfill(4)

#       Plak datum en tijd aan elkaar en converteer naar datetime object (format komt overeen met dftot)
dftot = df.datum + df.tijd
datum = pd.to_datetime(dftot, format='%Y%m%d%H%M')
```

```python
import datetime as dt
```

```python
#       Definieer tijdas met data en bepaal lengte van simulatie
delta       = dt.timedelta(minutes=10.)
delta2      = dt.timedelta(minutes=9.)
dates  = drange(datum[0], datum[len(df)-1]+delta2, delta)
simlen = len(dates)

#       Selecteer welke locaties we willen zien en over welke tijdstippen
locaties = stations[2:len(stations)]  # Haal namen van locaties op, behalve 1e 2 kolommen (datum/tijd)
t_start  = 0
t_eind   = len(dates)
tijdas   = dates[t_start:t_eind]
```
