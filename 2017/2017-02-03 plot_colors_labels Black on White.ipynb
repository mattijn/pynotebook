```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
```

```python
data = np.clip(randn(250,250),-1,1)
data = np.ma.masked_where(data>0.0,data)

fg_color = 'white'
bg_color = 'black'
fig, ax1 = plt.subplots(1,1, figsize=(7,5))

im = ax1.imshow(data, interpolation='nearest')
ax1.patch.set_facecolor(bg_color)
cb = plt.colorbar(im)

# IMSHOW    
# set title plus title color
ax1.set_title('ax1 title', color=fg_color)

# set tick and ticklabel color
im.axes.tick_params(color=fg_color, labelcolor=fg_color)

# set imshow outline
for spine in im.axes.spines.values():
    spine.set_edgecolor(fg_color)    

# COLORBAR
# set colorbar label plus label color
cb.set_label('colorbar label', color=fg_color)

# set colorbar tick color
cb.ax.yaxis.set_tick_params(color=fg_color)

# set colorbar edgecolor 
cb.outline.set_edgecolor(fg_color)

# set colorbar ticklabels
plt.setp(plt.getp(cb.ax.axes, 'yticklabels'), color=fg_color)

fig.patch.set_facecolor(bg_color)    
plt.tight_layout()
#plt.savefig(r'C:\Projects\Pr\3492.10\BruinKeijman\png\tmp//temp3.png', dpi=200, facecolor=bg_color)
plt.show()
```

```python
#as function
def plot_colors_labels(ax, cb, im, im_title, cb_label, fg_color='white', bg_color='black'):    
    # IMSHOW    
    # set title plus title color
    ax.set_title(im_title, color=fg_color)
    
    # set figure facecolor
    ax.patch.set_facecolor(bg_color)
    
    # set tick and ticklabel color
    im.axes.tick_params(color=fg_color, labelcolor=fg_color)
    
    # set imshow outline
    for spine in im.axes.spines.values():
        spine.set_edgecolor(fg_color)    

    # COLORBAR
    # set colorbar label plus label color
    cb.set_label(cb_label, color=fg_color)
    
    # set colorbar tick color
    cb.ax.yaxis.set_tick_params(color=fg_color)
    
    # set colorbar edgecolor 
    cb.outline.set_edgecolor(fg_color)
    
    # set colorbar ticklabels
    plt.setp(plt.getp(cb.ax.axes, 'yticklabels'), color=fg_color)
```

```python
data = np.clip(randn(250,250),-1,1)
data = np.ma.masked_where(data>0.0,data)

fig, [ax1,ax2] = plt.subplots(1,2, figsize=(16,4))

im = ax1.imshow(data, interpolation='nearest')
cb = fig.colorbar(im,ax=ax1, format='%.1f')
plot_colors_labels(ax1, cb, im, 'Albedo', 'Refl. coef.')

im = ax2.imshow(data, interpolation='nearest')
cb = fig.colorbar(im,ax=ax2, format='%.1f')
plot_colors_labels(ax2, cb, im, 'Albedo', 'Refl. coef.')

fig.patch.set_facecolor('black') 
plt.tight_layout()

plt.show()
```
