```python
import pandas as pd
import geopandas as gpd
import numpy as np
from scipy import sparse
import scipy.sparse.linalg as sp_linalg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_shp = r'D:\Projects\RO\Jong HKV\Toeleveringen\DominikPaprotny\ExtremeDischargesEUCatchments\Stations_total_v3.shp'
file_csv = r'D:\Projects\RO\Jong HKV\Toeleveringen\DominikPaprotny\ExtremeDischargesEUCatchments\BN_data_full.csv'
```

```python
# read and parse data
df = pd.read_csv(file_csv,sep=';', decimal=',')
df.replace(np.inf, 0, inplace=True)
df.replace(np.nan, 0, inplace=True)

# exclude some columns
df2 = df[df.columns.difference(['OBJECTID', 'Station_ID','WSO1_ID','Catchment_ID','Indicator_of_daily_discharge_availability'])]
df2 = df.astype(np.float32)
```

```python
x = df["OBJECTID"] #get index level values
unique_idx_last = x[np.invert(x.duplicated(keep='last'))].index
unique_idx_firs = x[np.invert(x.duplicated(keep='first'))].index
```

```python
unique_idx = np.concatenate((np.array(unique_idx_last),np.array(unique_idx_firs)))
```

```python
df3 = df2.iloc[unique_idx]
```

```python
# standardize matrix
X = df3.as_matrix()
_mu = np.nanmean(X, axis=0)
_sigma = np.sqrt(np.nanmean((X - _mu) ** 2.0, axis=0))
X = (X - _mu) / _sigma
```

```python
# go back to pandas dataframe and drop NaN columns
df = pd.DataFrame(data=X, columns=df.columns)
df.dropna(axis=1, how='all', inplace=True)
X = df.as_matrix()
```

```python
# calculate eigenvalues and eigenvectors using the SVD method
k = X.shape[1]-1
u, s, v = sp_linalg.svds(A=X, k=k)#np.linalg.svd(X)
eigenvals = s ** 2 # np.sqrt(corresponding eigenvalue via covariance matrix) == singular value s
eigenvecs = v.T # vectors are 'column-wise' determined    

# sort and calculate relevance statistics by order of eigenvalues
indices = np.argsort(eigenvals)
indices = indices[::-1] # sort eigenvalues from high to low
eigenvals = eigenvals[indices]
eigenvecs = eigenvecs[:, indices]    

# select the eigenpairs for usage
# [0] correspond to the 1st principal component
# [1] correspond to the 2nd principal component
# [0,1,2] correspond to the 1st, 2nd and 3rd principal component
# [1,3] combination of the 2nd and 4th principal component
ncomp = np.arange(0,k,1)
eigenvals=eigenvals[ncomp]    
eigenvecs=eigenvecs[:,ncomp]  

# determine the arrays of the principal components
factors = X.dot(eigenvecs)   

# select correct order of components to return
loadings = eigenvecs # V
coeff = eigenvecs.T # V'   

# normalize the principal components
coeff = (coeff.T * np.sqrt(eigenvals)).T
factors = factors / np.sqrt(eigenvals)   
```

```python
# data muggling as preparation for the figure
tot = np.ceil(sum(eigenvals)).astype(int)
var_exp = np.array([(i / tot)*100 for i in eigenvals])
cum_var_exp = np.cumsum(var_exp)
eigenvals_norm = np.array([i * 1/(tot/X.shape[1]) for i in eigenvals])
tot_norm = np.ceil(sum(eigenvals_norm)).astype(int)
x_ = np.arange(1,int(tot_norm),1)
```

```python
fig, axL = plt.subplots()

axL.bar(x_,eigenvals_norm, width=1, zorder=10, color="#4C72B0", label='eigenvalues')
axL.set_xlim(1,9)
axL.set_ylim(0,tot_norm)
axL.set_xlabel('principal components')
axL.set_ylabel('eigenvalues')
axL.set_xticks(x_)

axR = axL.twinx()
axR.plot(x_,cum_var_exp,color='cyan',marker='o',ls='', label='cumulative variance')
axR.set_ylim(0,100)
axR.set_ylabel('explaining variance (%)')
axR.set_yticks(np.around(np.linspace(axR.get_yticks()[0],axR.get_yticks()[-1],len(axL.get_yticks())),decimals=2))
axR.grid(None)

# for i,label, in enumerate([str(np.round(i*100, 1)) for i in pca.rsquare[1:]]):    
#     z = cum_var_exp
#     axR.annotate(label,(x_[i]-0.05,z[i]-6))

# ask matplotlib for the plotted objects and their labels
lines, labels = axL.get_legend_handles_labels()
lines2, labels2 = axR.get_legend_handles_labels()
axL.legend(lines + lines2, labels + labels2, loc=4)

plt.title('scree plot + explaining variance')
plt.tight_layout()

plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0, color='orange')
    ax.annotate('', v1, v0, arrowprops=arrowprops)
```

```python
# plot principal components
X_pca = pca.transform(X)
sc = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['Annual_maxima_of_daily_discharge'],alpha=0.2, label='EU river catchments')
cbar = plt.colorbar(sc)
cbar.ax.set_ylabel('Annual_maxima_of_daily_discharge')
draw_vector([0, 0], [0, 5])
draw_vector([0, 0], [5, 0])
plt.legend()
plt.xlim(-15,15)
plt.ylim(-15,15)
#plt.axis('equal')

plt.xlabel('component 1')
plt.ylabel('component 2')
plt.title('principal components')
plt.show()
```
