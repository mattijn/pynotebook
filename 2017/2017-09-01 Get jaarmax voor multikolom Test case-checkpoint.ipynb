```python
import pandas as pd
import numpy as np

rng = pd.date_range('2010-1-1', periods=1000, freq='D')
df = pd.DataFrame(np.random.randn((len(rng)),2), index=rng, columns=['A','B'])
df.head()
```

```python
startMM, startdd = (2,15)
endMM, enddd = (10,3)
```

```python
subset = df[((df.index.month == startMM) & (startdd <= df.index.day) 
             | (df.index.month == endMM) & (df.index.day <= enddd))]
```

```python
# groupy year en selecteer de max. Returned volle datums waar de max van dat jaar was 
idx = df.groupby([df.index.year])['A'].transform(max) == df['A']
df[idx]
```
