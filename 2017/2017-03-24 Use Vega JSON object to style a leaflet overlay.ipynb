```python
import folium
import geopandas as gpd
import pandas as pd
```

```python
json_nl = r'D:\Projects\RO\HydroAsAService\gemeentegrenzen_topo.json'
```

```python
gdf = gpd.read_file(json_nl)
# convert shape to something more normal number
gdf['SHAPE_Area'] /= 10e7
gdf.head()
```

```python
map_flevoland = folium.Map(location=[52.506867, 5.469034],tiles='Stamen Toner')
folium.Marker([52.506867, 5.469034], popup='Lelystad').add_to(map_flevoland)
folium.Marker([52.374112, 5.214609], popup='Almere').add_to(map_flevoland)

#map_flevoland.save('map_flevoland.html')
map_flevoland.choropleth(geo_path=json_nl, topojson='objects.gemeentegrenzen_geo',
                         data=gdf,columns=['GM_CODE','SHAPE_Area'], key_on='feature.properties.GM_CODE',
                         fill_color='YlGnBu', line_opacity=0.3,legend_name='Oppervlak gemeente',highlight=True
                        
                        )
map_flevoland
```

```python
import folium
import pandas as pd

state_geo = r'D:\Projects\RO\HydroAsAService\us_counties_20m_topo.json'
state_unemployment = r'D:\Projects\RO\HydroAsAService\US_Unemployment_Oct2012.csv'

state_data = pd.read_csv(state_unemployment)

#Let Folium determine the scale
map = folium.Map(location=[48, -102], zoom_start=3)
map.choropleth(geo_path=state_geo, data=state_data,
             columns=['State', 'Unemployment'],
             key_on='feature.id',
             fill_color='YlGn', fill_opacity=0.7, line_opacity=0.2,
             legend_name='Unemployment Rate (%)')
#map.save('us_states.html')
map
```
