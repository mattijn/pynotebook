```python
from osgeo import gdal
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
msk_path = r'G:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\2009\OUTPUT_SummerOnly\COMB//psdx.tif'
msk_ds = gdal.Open(msk_path)
msk_bd = msk_ds.GetRasterBand(1)
msk_ar = msk_bd.ReadAsArray()

path = r'C:\Users\Matt\Downloads\AWC_CLASS2\phase//ph_2009.tif'
ds = gdal.Open(path)
bd = ds.GetRasterBand(1)
ar = bd.ReadAsArray()

#ph08_msk = np.ma.masked_where(msk_ar <= msk_ar[0,0], ar)
ph09_msk = np.ma.masked_where(msk_ar <= msk_ar[0,0], ar)

#im_ph08 = plt.imshow(ph08_msk)
im_ph09 = plt.imshow(ph09_msk)
#plt.colorbar(im_ph08)
plt.colorbar(im_ph09)
plt.show()
```

```python
path = r'C:\Users\Matt\Downloads\AWC_CLASS2\meris//meris_cn.tif'
ds = gdal.Open(path)
bd = ds.GetRasterBand(1)
ar = bd.ReadAsArray()

meris_msk = np.ma.masked_where(msk_ar <= msk_ar[0,0], ar)

im_meris = plt.imshow(meris_msk)
plt.colorbar(im_meris)
plt.show()
```
