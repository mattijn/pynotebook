```python
import mpl_toolkits
from mpl_toolkits.basemap import Basemap 
import matplotlib.pyplot as plt 
import numpy as np
from osgeo import gdal
import os
```

```python
def rasterList(varpath, endwith, namefile):
    # 1 - Precipitation
    input_folder = varpath
    varList = [os.path.join(root, name)
                 for root, dirs, files in os.walk(input_folder)
                 for name in files
                 if name.endswith((endwith))]
    print namefile + " files : "  + str(len(varList))
    return varList
```

```python
fils = rasterList(r'I:\Data\anomaly','tif','ano_doys')
```

```python
def openRaster(pathToRaster):
    #pathToRaster = r'I:\Data\anomaly//ano_DOY2002170.tif'
    raster = gdal.Open(pathToRaster, gdal.GA_ReadOnly)
    array = raster.GetRasterBand(1).ReadAsArray()
    msk_array = np.ma.masked_equal(array, value = 65535)
    array = None
    return raster, msk_array
```
