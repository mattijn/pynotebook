```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
def project1(projections):
    m = Basemap(projection=projections,resolution='c',lat_0=0,lon_0=0,
                boundinglat=0)#lat_0=41.508577,lon_0=148.331909,)#
    # draw parallels and meridians.
    m.drawmeridians(np.arange(0,360,30)) 
    m.drawparallels(np.arange(-90,90,30)) 
    
    m.drawcoastlines(linewidth=0.5) 
    #m.drawmapboundary()
    m.drawcountries()

    plt.title(projections+' projection')
    plt.show()
list1 = ['ortho', 'vandg', 'nsper', 'aeqd', 'merc','cyl', 'mill', 
         'gall', 'robin','mbtfpq','sinu', 'moll', 'hammer', 
         'npstere', 'spstere', 'npaeqd',] #17          
for pr in list1:
    print pr
    project1(pr)
```
