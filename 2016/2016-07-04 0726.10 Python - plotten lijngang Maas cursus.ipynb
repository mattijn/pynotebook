```python
Maas = pd.read_csv('Maas.csv', sep=';', header=None)
Maas.columns = ['x', 'y']
Maasxy = np.c_[Maas.x.values[:], Maas.y.values[:]]
```

```python
shp.LineString(Maasxy)
```

```python
Maas_ = shp.LineString(Maasxy)
```

```python
waterstanden = pd.read_csv('waterstanden.csv', sep=';', header=None)
waterstanden.columns = ['km', 'x', 'y', 'ws']
waterstanden.head()
```

```python
data = pd.read_csv(r'waterstanden.csv', header = None, sep=';').values[:]
```

```python
chainage = np.zeros(len(data))
```

```python
for i, d in enumerate(data):
    pt = shp.Point(d[1],d[2])
    dist = Maas_.project(pt)
    chainage[i] = dist
```

```python
chainage
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(chainage,waterstanden.ws)
```
