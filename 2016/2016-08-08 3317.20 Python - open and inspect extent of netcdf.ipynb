```python
from netCDF4 import Dataset
import numpy as np
from cartopy import config
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
my_example_nc_file = r'D:\Users\hoek\Downloads//output (7).nc'
fh = Dataset(my_example_nc_file, mode='r')
```

```python
lons = fh.variables['x'][:]
lats = fh.variables['y'][:]
ws = fh.variables['VELV'][:]

ws_units = fh.variables['VELV'].units

lon_0 = lons.mean()
lat_0 = lats.mean()
```

```python
l_n, l_t = np.meshgrid(lons,lats)
```

```python
plt.figure(figsize=(20,10))
ax = plt.axes(projection=ccrs.Stereographic())

#plt.contourf(lons, lats, ws[0][0], transform=ccrs.PlateCarree())
plt.pcolormesh(l_n, l_t, ws[0][0])
#ax.coastlines()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Generate some data
x, y, z = np.random.random((3, 30))
z = z * 20 + 0.1

# Set some values in z to 0...
z[:5] = 0

cmap = plt.get_cmap('viridis', 13)
cmap.set_under('gray')

fig, ax = plt.subplots(figsize=(10,10))

cax = ax.scatter(x, y, c=z, s=100, cmap=cmap, vmin=0.1, vmax=z.max())
fig.colorbar(cax, extend='min')

plt.show()
```
