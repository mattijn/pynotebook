```python
from xml.etree import ElementTree
from xml.dom import minidom
import glob, os
from xml.etree.ElementTree import Element, SubElement, Comment
import subprocess as sp
import pandas as pd
from datetime import datetime
```

```python
directory = r'D:\Projects\Pr\3317.20\Toeleveringen\divide'
asc_dir = r'D:\Projects\Pr\3317.20\Toeleveringen\ASC_out_inc_noDataVal'
tif_dir = r'D:\Projects\Pr\3317.20\Toeleveringen\TIF_out_inc_noDataVal'
gdal_translate = r'C:\Python35\Lib\site-packages\osgeo//gdal_translate.exe'
gdalwarp = r'C:\Python35\Lib\site-packages\osgeo//gdalwarp.exe'
```

```python
for i in range(0,25):
    directory = r'D:\Projects\Pr\3317.20\Toeleveringen\divide\vak'+str(i)

    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.tif'):
                old_name = os.path.join( os.path.abspath(root), file )
                base, extension = os.path.splitext(file)
                print ('\n'+base)
                
                # set nodata value in tif                
                tif_out = os.path.join(tif_dir,base+'.tif')
                command = [gdalwarp,'-dstnodata', '0','-overwrite',old_name,tif_out]
                print (sp.list2cmdline(command))
                norm = sp.Popen(sp.list2cmdline(command),stdout=sp.PIPE, shell=True)
                norm.communicate()                
                print ('\n')
                # convert tif to asc
                asc_out = os.path.join(asc_dir,base+'.asc')
                command = [gdal_translate,'-of','AAIGrid',tif_out,asc_out]
                print (sp.list2cmdline(command))
                norm = sp.Popen(sp.list2cmdline(command),stdout=sp.PIPE, shell=True)
                norm.communicate()
```
