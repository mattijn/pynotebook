```python
import pandas as pd
import numpy as np
```

```python
code = []
omschrijving = []
with open(r'D:\Projects\RO\0815.10 Data Challenge 2018\N_S_WSS\standaard_schadetabel.txt') as f: 
    for idx, line in enumerate(f): 
        if line.startswith('['):
            print(idx, line)
            code.append(line)
        if line.startswith('omschrijving'):
            print(idx, line)
            omschrijving.append(line)            
```

```python
code = [c.lstrip('[') for c in code]
code = [c.strip('\n') for c in code]
code = [c.rstrip(']') for c in code]
code = [int(c) for c in code]
```

```python
omschrijving = [oms.rstrip('\n') for oms in omschrijving]
omschrijving = [oms.lstrip('omschrijving = ') for oms in omschrijving]
```

```python
df = pd.DataFrame(data=np.array([omschrijving,code]).T, columns = ['omschrijving', 'code'])
```

```python
df.to_csv(r'D:\Projects\RO\0815.10 Data Challenge 2018//code_omschrijving.csv')
```
