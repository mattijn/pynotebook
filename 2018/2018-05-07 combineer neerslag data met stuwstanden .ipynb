```python
import pandas as pd
```

```python
neerslag_csv = r'D:\Projects\RO\0815.10 Data Challenge 2018\Neerslag\neerslag_mm_hr.csv'
```

```python
df = pd.read_csv(neerslag_csv)
df.set_index('datetime', inplace=True)
df.index = pd.to_datetime(df.index)
df.columns = ['peilvak.{}'.format(ix) for ix in [132, 15, 34, 95, 92, 21, 113, 65, 142, 12, 100, 74]]
df.head()
```

```python
stuw_csv = r'D:\Projects\RO\0815.10 Data Challenge 2018\Neerslag\stuw_standen_peilvak.csv'
df_stuw = pd.read_csv(stuw_csv)
df_stuw.set_index('datetime', inplace=True)
df_stuw.index = pd.to_datetime(df_stuw.index)
df_stuw = df_stuw.resample('H').mean()
df_stuw.head()
```

```python
df_1H = df.groupby(pd.TimeGrouper('H')).sum()
df_2H = df_1H.rolling(window = '2H').sum()
df_3H = df_1H.rolling(window = '3H').sum()
df_4H = df_1H.rolling(window = '4H').sum()
df_5H = df_1H.rolling(window = '5H').sum()
df_6H = df_1H.rolling(window = '6H').sum()
df_7H = df_1H.rolling(window = '7H').sum()
df_8H = df_1H.rolling(window = '8H').sum()
df_9H = df_1H.rolling(window = '9H').sum()
df_10H = df_1H.rolling(window = '10H').sum()
df_11H = df_1H.rolling(window = '11H').sum()
df_12H = df_1H.rolling(window = '12H').sum()
df_13H = df_1H.rolling(window = '13H').sum()
df_14H = df_1H.rolling(window = '14H').sum()
df_15H = df_1H.rolling(window = '15H').sum()
df_16H = df_1H.rolling(window = '16H').sum()
df_17H = df_1H.rolling(window = '17H').sum()
df_18H = df_1H.rolling(window = '18H').sum()
df_19H = df_1H.rolling(window = '19H').sum()
df_20H = df_1H.rolling(window = '20H').sum()
df_21H = df_1H.rolling(window = '21H').sum()
df_22H = df_1H.rolling(window = '22H').sum()
df_23H = df_1H.rolling(window = '23H').sum()
df_24H = df_1H.rolling(window = '24H').sum()
df_32H = df_1H.rolling(window = '32H').sum()

df_1H_FC = df_1H.shift(freq= pd.Timedelta('-1 hour'))
df_2H_FC = df_2H.shift(freq= pd.Timedelta('-2 hour'))
df_3H_FC = df_3H.shift(freq= pd.Timedelta('-3 hour'))
df_4H_FC = df_4H.shift(freq= pd.Timedelta('-4 hour'))
df_5H_FC = df_5H.shift(freq= pd.Timedelta('-5 hour'))
df_6H_FC = df_6H.shift(freq= pd.Timedelta('-6 hour')) 

from_date = '2015-08-30 18:00'
to_date = '2015-08-31 08:00'


for peilvak in df_stuw.columns:
    print(peilvak)
    
    comb_peilvak = pd.concat([df_1H[peilvak], df_2H[peilvak], df_4H[peilvak], df_8H[peilvak], df_16H[peilvak], 
                              df_32H[peilvak], df_1H_FC[peilvak], df_2H_FC[peilvak], df_3H_FC[peilvak], 
                              df_4H_FC[peilvak], df_5H_FC[peilvak], df_6H_FC[peilvak]], axis=1)
    comb_peilvak.columns = ['X.last1Hsum', 'X.last2Hsum', 'X.last4Hsum', 'X.last8Hsum', 'X.last16Hsum', 
                            'X.last32Hsum', 'X.coming1Hsum', 'X.coming2Hsum', 'X.coming3Hsum', 
                            'X.coming4Hsum', 'X.coming5Hsum', 'X.coming6Hsum']
    
    df_Y_waterlevel = df_stuw[peilvak]
    df_Y_waterlevel.rename('Y.waterlevel', inplace=True)
    df_comb = pd.concat((comb_peilvak, df_Y_waterlevel), axis=1, join='inner')    
    
    pv_out_csv = r'D:\Projects\RO\0815.10 Data Challenge 2018\Neerslag//__{}.csv'.format(peilvak)
    df_comb.to_csv(pv_out_csv)    
```
