```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_pct_schade(schade_pct):
    fig = plt.figure(figsize=(2,3))
    
    plt.yscale('log')
    #ax1 = fig.add_subplot(111)
    bar_max, = plt.bar(range(1, 2), 100, bottom=0.1, color='lightgray', edgecolor='black', alpha=0.5)
    bar_current, =plt.bar(range(1, 2), schade_pct, bottom=0.1, color='cyan', edgecolor='black')
    plt.xticks([1.5])
    plt.yticks([0.1,1,10,100], [0.1,1,10,100])
    plt.ylabel('schade (%)')

    ax = fig.axes[0]
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    

    
    # Shrink current axis by 20%
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

    # Put a legend to the right of the current axis
    ax.legend((bar_max, bar_current), ('maximale schade', 'potentiele schade ({}%)'.format(schade_pct)), loc='center left', bbox_to_anchor=(1, 0.5)).get_frame().set_linewidth(0.0)
    ax.set_title('datum hier')
    plt.show()
```

```python
maai_csv = r'D:\Projects\RO\0815.10 Data Challenge 2018\maaiveldcurves\BigdataMaaivelcurve_gesorteerd.xls'
```

```python
df = pd.read_excel(maai_csv)
```

```python
peilvak= 95
df_sel = df[df['GEBIEDID']==peilvak]
ax = df_sel.plot('PERCENTAGE', 'MAAIVELD', legend=False, title='geinundeerd gebied (%) van peilvak {}'.format(peilvak))
ax.set_ylabel('hoogte in mNAP')
ax.set_xlabel('percentage in %')
plt.show()
```

```python
hoogte_mNAP = [-1.5, -1, 0.5, 1, 2]
```

```python
for h_mNAP in hoogte_mNAP:
    df_nearest = df.iloc[(df['MAAIVELD']-h_mNAP).abs().argsort()][0:1][['MAAIVELD','PERCENTAGE']]
    schade_pct = df_nearest['PERCENTAGE'].values[0]
    print(schade_pct)
    plot_pct_schade(schade_pct)
```
