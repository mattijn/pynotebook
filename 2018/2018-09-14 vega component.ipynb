```python
import altair as alt
import pandas as pd
```

```python
df_Q = pd.read_csv(r"LSHM_Q_1271500142.csv")
df_Q['Date'] = df_Q['Date'].str[1:-1]
```

```python
df_h = pd.read_csv(r"H_1271500142.csv")
```

```python
df_h.head()
```

```python
df_Q.head()
```

```python
df_h['date'][0]
```

```python
df_Q_h = df_h.merge(df_Q, how='outer', left_on='date', right_on='Date')
df_Q_h['Date'] = pd.to_datetime(df_Q_h['Date'])
df_Q_h.set_index('Date', inplace=True)
df_Q_h = df_Q_h['2018-06-01':'2018-09-18'].reset_index()
df_Q_h['His10'] = df_Q_h['His10'] 
df_Q_h['His90'] = df_Q_h['His90'] 
```

```python
df_Q_h.head()
```

```python
line_his = alt.Chart(df_Q_h).mark_line(color='blue').encode(
    x='Date:T',
    y='Qref'
)

point_his = alt.Chart(df_Q_h).mark_point(color='blue').encode(
    x='Date:T',
    y='Qref',
    tooltip=['Qref']
)

line_fc = alt.Chart(df_Q_h).mark_line(color='red').encode(
    x='Date:T',
    y='Qforecast'
)

point_fc = alt.Chart(df_Q_h).mark_point(color='red').encode(
    x='Date:T',
    y='Qforecast',
    tooltip=['Qforecast']
)


area_hist = alt.Chart(df_Q_h).mark_area(color='lightblue', opacity=0.4).encode(
    x='Date:T',
    y='His10',
    y2='His90'
).interactive(bind_y=False)

area_hist+line_his+line_fc+point_his+point_fc
```

```python
df_wl = pd.read_csv(r"WaterLevelAlerts.csv")
df_wl = df_wl[df_wl['Station']==1271500142]
df_wl.head()
```

```python
df_Q_h['Red'] = df_wl['Red'].tolist()[0]
df_Q_h['Orange'] = df_wl['Orange'].tolist()[0]
df_Q_h['Yellow'] = df_wl['Yellow'].tolist()[0]
df_Q_h['Green'] = df_wl['Green'].tolist()[0]
```

```python
line_his = alt.Chart(df_Q_h).mark_line(color='blue').encode(
    x='Date:T',
    y='waterLevel'
)

point_his = alt.Chart(df_Q_h).mark_point(color='blue').encode(
    x='Date:T',
    y='waterLevel',
    tooltip=['waterLevel']
)

line_fc = alt.Chart(df_Q_h).mark_line(color='red').encode(
    x='Date:T',
    y='Hforecast'
)

point_fc = alt.Chart(df_Q_h).mark_point(color='red').encode(
    x='Date:T',
    y='Hforecast',
    tooltip=['Hforecast']
).interactive(bind_y=False)

rule_red = alt.Chart(df_wl).mark_rule(color='purple').encode(
    y='Orange:Q'
)

rule_orange = alt.Chart(df_wl).mark_rule(color='red').encode(
    y='Yellow:Q'
)

rule_yellow = alt.Chart(df_wl).mark_rule(color='orange').encode(
    y='Green:Q'
)

# rule_green = alt.Chart(df_wl).mark_rule(color='green').encode(
#     y='Green:Q'
# )

line_his+line_fc+point_his+point_fc+rule_red+rule_orange+rule_yellow#+rule_green
```

```python
data = pd.DataFrame({"Day": range(1, 16),
                     "Value": [54.8, 112.1, 63.6, 37.6, 79.7, 137.9, 120.1, 103.3,
                               394.8, 199.5, 72.3, 51.1, 112.0, 174.5, 130.5]})

data2 = pd.DataFrame([{"ThresholdValue": 300, "Threshold": "hazardous"}])

bar1 = alt.Chart(data).mark_bar().encode(
    x='Day:O',
    y='Value:Q'
)

bar2 = alt.Chart(data).mark_bar(color="#e45755").encode(
    x='Day:O',
    y='baseline:Q',
    y2='Value:Q'
).transform_filter(
    "datum.Value >= 300"
).transform_calculate(
    "baseline", "300"
)

rule = alt.Chart(data2).mark_rule().encode(
    y='ThresholdValue:Q'
)

text = alt.Chart(data2).mark_text(
    align='left', dx=215, dy=-5
).encode(
    alt.Y('ThresholdValue:Q', axis=alt.Axis(title='PM2.5 Value')),
    text=alt.value('hazardous')
)

bar1 + text + bar2 + rule
```
