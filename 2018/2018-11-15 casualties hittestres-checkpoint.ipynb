```python
import cbsodata
import knmi
import pandas as pd
import altair as alt
from datetime import datetime
from scipy.signal import savgol_filter
%matplotlib inline
```

```python
def period_yearweek(period):
    """
    Function to parse period object to '%Y-%U' object
    """
    list_week = period.split(' ')
    year = list_week[0].zfill(4)
    week = list_week[2].zfill(2)
    year_week = year+'-'+week
    return year_week
```

```python
tables = pd.DataFrame(cbsodata.get_table_list())
```

```python
# get data by identifier from CBR Open Data
df_cs_week = pd.DataFrame(cbsodata.get_data('70895ned'))
```

```python
# filter dataframe to get all males and females for all age-groups
df_mf_allage = df_cs_week[(df_cs_week['Geslacht'] =='Totaal mannen en vrouwen') & (df_cs_week['LeeftijdOp31December'] =='Totaal leeftijd')]
df_all_weeks = df_mf_allage[df_mf_allage['Perioden'].str.contains('week')]
```

```python
# parse periods to datetime objects
df_all_weeks.loc[:,'yearweek'] = [period_yearweek(period) for period in df_all_weeks['Perioden']]
df_all_weeks.loc[:,'yearweek'] = pd.to_datetime(df_all_weeks.yearweek.add('-0'), format='%Y-%W-%w')
df_all_weeks.loc[:,'year'] = df_all_weeks.yearweek.dt.strftime('%Y')
df_all_weeks.loc[:,'week'] = df_all_weeks.yearweek.dt.strftime('%W')
df_all_weeks.year = df_all_weeks.year.astype(int)
df_all_weeks.week = df_all_weeks.week.astype(int)
```

```python
# group by week and get mean
s_cs_week = df_all_weeks.groupby('week')['Overledenen_1'].mean()
cs_week_mean = s_cs_week.reset_index()
```

```python
df_all_weeks = df_all_weeks.join(cs_week_mean[2:-1], on='week', how='left', rsuffix='_mean')
```

```python
df_all_weeks.head()
```

```python
s_anomaly = df_all_weeks['Overledenen_1'] - df_all_weeks['Overledenen_1_mean']
s_anomaly.index = df_all_weeks.yearweek
s_anomaly.name = 'delta_casualties'
s_anomaly.interpolate(inplace=True)
```

```python
s_anomaly.tail()
```

```python
alt.Chart(s_anomaly.reset_index()).mark_line().encode(
    x='yearweek',
    y='delta_casualties'
).interactive(bind_y=False)
```

```python
# get data for station de bilt (260)
df_ug_tg_day = knmi.get_day_data_dataframe([260],  datetime(1995,1, 1), datetime(2018,10, 28), inseason=False, variables=['UN','UG','UX','TN','TG','TX'])
df_ug_tg_day = df_ug_tg_day.tz_localize(None)
```

```python
# resample to week mean values and convert units
df_ug_tg_week = df_ug_tg_day.resample('W', how='mean')
df_ug_tg_week = df_ug_tg_week.reset_index()[['YYYYMMDD','UN','UG','UX','TN','TG','TX']]
df_ug_tg_week[['TN','TG','TX']] = df_ug_tg_week[['TN','TG','TX']] * 0.1
```

```python
# add SG filtered series
df_ug_tg_week.loc[:,'UN_SG9'] = savgol_filter(df_ug_tg_week['UN'], window_length=9, polyorder=2)
df_ug_tg_week.loc[:,'UG_SG9'] = savgol_filter(df_ug_tg_week['UG'], window_length=9, polyorder=2)
df_ug_tg_week.loc[:,'UX_SG9'] = savgol_filter(df_ug_tg_week['UX'], window_length=9, polyorder=2)
df_ug_tg_week.loc[:,'TN_SG9'] = savgol_filter(df_ug_tg_week['TN'], window_length=9, polyorder=2)
df_ug_tg_week.loc[:,'TG_SG9'] = savgol_filter(df_ug_tg_week['TG'], window_length=9, polyorder=2)
df_ug_tg_week.loc[:,'TX_SG9'] = savgol_filter(df_ug_tg_week['TX'], window_length=9, polyorder=2)
```

```python
# add week and year column
df_ug_tg_week.loc[:,'year'] = df_ug_tg_week.YYYYMMDD.dt.strftime('%Y')
df_ug_tg_week.loc[:,'week'] = df_ug_tg_week.YYYYMMDD.dt.strftime('%W')
df_ug_tg_week.year = df_ug_tg_week.year.astype(int)
df_ug_tg_week.week = df_ug_tg_week.week.astype(int)
```

```python
df_ug_tg_week.head()
```

```python
# group by week and get mean
df_un_tg_week_mean = df_ug_tg_week.groupby('week')[['UN','UG','UX','TN','TG','TX','UN_SG9','UG_SG9','UX_SG9','TN_SG9','TG_SG9','TX_SG9']].mean()
df_un_tg_week_mean = df_un_tg_week_mean.reset_index()
```

```python
# join week means on orginal values
df_ug_tg_week = df_ug_tg_week.join(df_un_tg_week_mean, on='week', how='left', rsuffix='_mean')
```

```python
df_ug_tg_week.head()
```

```python
# compute anomalies
col_couples = [['UN','UN_mean'],['UG','UG_mean'],['UX','UX_mean'],['TN','TN_mean'],['TG','TG_mean'],['TX','TX_mean'],
               ['UN_SG9','UN_SG9_mean'],['UG_SG9','UG_SG9_mean'],['UX_SG9','UX_SG9_mean'],['TN_SG9','TN_SG9_mean'],
               ['TG_SG9','TG_SG9_mean'],['TX_SG9','TX_SG9_mean']
              ]
df_anomaly = pd.DataFrame()
for couple in col_couples:
    anomaly = df_ug_tg_week[couple[0]] - df_ug_tg_week[couple[1]]
    anomaly.index = df_ug_tg_week.YYYYMMDD
    anomaly.name = 'delta_{}'.format(couple[0])
    df_anomaly = pd.concat((df_anomaly, anomaly), axis=1)
#df_anomaly = df_anomaly.tz_localize(None)
```
