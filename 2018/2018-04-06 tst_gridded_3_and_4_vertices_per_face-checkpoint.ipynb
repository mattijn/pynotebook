For the mesh, we get multiple variables from the grid object. Namely, the `node_lon`, `node_lat`, `faces`, `edges` and `nodes`.
- `node_lon` is 1D array with longitude values
- `node_lat` is 1D array with latitude values
- `faces` is 2D array, where each row contains the vertices of the face/patch
- `edges` is 2D array, where each row contains the first vertice of the  the face/patch
- `nodes` is 2D array, where each row contains the coordinate corresponding the vertice

```python
# note that we are working with just the variable object now...
lon = var.grid.node_lon
lat = var.grid.node_lat
faces = var.grid.faces
edges = var.grid.edges
nodes = var.grid.nodes
```

```python
# convert the squares into `Polygon` patches and add to `PatchCollection` 

# Note that we use "gridcells2"

patches = []
for gridcell in tqdm(gridcells2):    
    polygon = Polygon(gridcell, closed=True)
    patches.append(polygon)      

collection = PatchCollection(patches, cmap='viridis', alpha=1, zorder=10, edgecolor="")

# now there is a collection containing the patches, but each patch has not yet a fill value.
# lets use the `data_coloring`
collection.set_array(data_coloring)

print(len(patches))
```

Prepare faces for matplotlib, this means the defining function for faces

```python
def faces_plot(x,y, faces, colors, ax=None, **kwargs):
    if not ax: ax=plt.gca()
    xy = np.c_[x,y]
    #verts  = xy[faces]
    # BELOW to account for triangles in grid
    verts2 = []
    for face in faces :
        if face.min()>0 :
            #print(face) 
            verts2.append(xy[face])
        else : # than triangles
            verts2.append(xy[face[0:3]])
    pc = PolyCollection(verts2, **kwargs)
    pc.set_array(colors)
    ax.add_collection(pc)
    ax.autoscale()
    return pc
```

And plot using cartopy

```python
def make_map(lon, lat, projection=ccrs.PlateCarree()):
    fig, ax = plt.subplots(figsize=(15,8),
                           subplot_kw=dict(projection=projection))
    
    ax.set_extent([lon.min(), lon.max(), lat.min(), lat.max()])

    ax.coastlines(resolution='10m')
    
    gl = ax.gridlines(draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')
    gl.xlabels_top = gl.ylabels_right = False
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    return fig, ax
```

```python
# create a map
#fig, ax = make_map(lat, lon, projection=ccrs.epsg(28992))
fig, ax = make_map(lon, lat)

# add patches and set extent
ax.add_collection(collection)
pc = faces_plot(lon, lat, faces, data_coloring, ax=None, cmap='viridis', color="", facecolor="None")

# prepare a location for a colorbar
divider = make_axes_locatable(ax)
ax_cb = divider.new_horizontal(size="5%", pad=0.1, axes_class=plt.Axes)

# add the figure and colorbar
fig.add_axes(ax_cb)
cbar = plt.colorbar(pc, cax=ax_cb)

ax.set_title('{} ({})'.format(
    var.attributes['long_name'], time_interest.strftime("%B %d, %Y")))
cbar.ax.set_ylabel(var.attributes['units'])

plt.show()
```
