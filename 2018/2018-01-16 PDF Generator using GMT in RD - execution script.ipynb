```python
import io
import nbformat

def execute_notebook(nbfile):
    with io.open(nbfile, encoding="utf8") as f:
        nb = nbformat.read(f, as_version=4)

    ip = get_ipython()

    for cell in nb.cells:
        if cell.cell_type != 'code':
            continue
        ip.run_cell(cell.source)
```

```python
# collect new source data files
new_observations = glob.glob(os.path.join(ruweDataDir, '*.pts'))[11:12]
print('no. of files: {}'.format(len(new_observations)))
print(new_observations)
```

```python
i = 0
for file in tqdm(new_observations):    
    # only select multibeam (not SIlAS) .pts files 
    if file.endswith('.pts') and r'SILAS' not in file:
        print('file: {0} / {1}\n'.format(i, file))
        
        # convert pts file to GMT nc-file and get geodataframe of polygon features containing x% points
        gmt_object = pts2gmtnc_idxlist(
            file, ruweDataDir, tmpDataDir, gdf, idx=i, rd_string=rd_string)        

        # create location list of baggervakken to plot
        idx_list = []
        for idx, item in enumerate(gmt_object['gdf_bgvk_sel'].OBJECTNAAM):
            print('index loc: {0} / named: {1}'.format(idx, item))
            
            # get the location index of each file
            loc_feature = gdf[gdf.OBJECTNAAM == item].index.tolist()[0]
            idx_list.append(loc_feature)
            
        print('\nfile contains: {} dregdefield(s) to parse into PDF\n'.format(len(idx_list)))

        # iterate over the polygon features and plot using GMT
        for idx in tqdm(idx_list):            
            # get the plot parameters for bounding box
            gmt_extent = mmbox(gdf, idx)
            
            # get the plot parameters for scalebar
            gmt_extent = get_scalebar_params(gdf, idx, gmt_extent, gmt_object)
            
            # get name baggervak            
            out_name = 'DATUM-{0}-BVAK-{1}.ps'.format(gmt_object['file_gmt_date'].strftime(
                '%Y%m%d'), urlify(gmt_extent['title'].replace('.', '_')))
            
            # set path to PS
            gmt_object['file_out_ps'] = os.path.join(
                outDataDir, out_name)         
            
            # set path to betonningen
            gmt_object['file_gmt_ton_color'] = r'D:\Projects\Pr\3317.20\PTS2PDF_usingGMT\data_dir_ton\TON_07sep2017_ton_color.txt'
            gmt_object['file_gmt_ton_label'] = r'D:\Projects\Pr\3317.20\PTS2PDF_usingGMT\data_dir_ton\TON_07sep2017_ton_label.txt'
            gmt_object['file_gmt_baggervak_lon_lat'] = r'D:\Projects\Pr\3317.20\PTS2PDF_usingGMT\data_dir_tmp\location_label.txt'
            
            # set path to colorbar
            gmt_object['custom_cpt'] = os.path.join(
                tmpDataDir, 'CB_{}.cpt'.format(urlify(gmt_extent['title'].replace('.', '_'))))            
            
            # get the plot paramters for stamp
            gmt_stamp = set_gmt_stamp(gmt_object, gmt_extent, logosDtDir, tmpDataDir, idx=i)
            create_stamp(gmt_stamp, show_not_save=False)
            
            # set the stamp location in gmt_object
            gmt_object['file_stamp'] = gmt_stamp['path_to_stamp']

            # do the plotting
            apply_gmt(gmt, gmt_object, gmt_extent)
        
        # finished parsing .pts file. Add 1!
        i += 1

print('total files: {}'.format(i))
```
