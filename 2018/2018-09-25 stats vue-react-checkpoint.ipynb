```python
df_npm_github = pd.concat((df_melt_npm, df_melt_github), axis=0).set_index('index').sort_index()
df_npm_github.head()
```

```python
# interpolate github
df_react_tmp = df_npm_github[df_npm_github['library'] == 'react']['stars-on-github']
df_npm_github.loc[df_npm_github['library'] == 'react', 'stars-on-github'] = df_react_tmp.interpolate().reset_index()['stars-on-github'].tolist()

df_vue_tmp = df_npm_github[df_npm_github['library'] == 'vue-js']['stars-on-github']
df_npm_github.loc[df_npm_github['library'] == 'vue-js', 'stars-on-github'] = df_vue_tmp.interpolate().reset_index()['stars-on-github'].tolist()

# # interpolate npm
df_react_tmp = df_npm_github[df_npm_github['library'] == 'react']['npm-installs']
df_npm_github.loc[df_npm_github['library'] == 'react', 'npm-installs'] = df_react_tmp.interpolate().reset_index()['npm-installs'].tolist()

df_vue_tmp = df_npm_github[df_npm_github['library'] == 'vue-js']['npm-installs']
df_npm_github.loc[df_npm_github['library'] == 'vue-js', 'npm-installs'] = df_vue_tmp.interpolate().reset_index()['npm-installs'].tolist()
```

```python
df_npm_github.reset_index(inplace=True)
```
