```python
import subprocess as sp
import glob
import os
import concurrent.futures
```

```python
def cmd(command, parse_directly=False):
    
    if parse_directly==True:
        #print(command)
        norm = sp.Popen(command,stdout=sp.PIPE, shell=True)
    else:
        #print(sp.list2cmdline(command))
        norm = sp.Popen(sp.list2cmdline(command),stdout=sp.PIPE, shell=True)            
             
    out_cmd = norm.communicate()
    return out_cmd
```

```python
def func_fillnodata(in_file):
    # The fill_nodata will be named "<original_filename>_fillnodata.asc"
    base_filename, file_extension = os.path.splitext(in_file)
    tmp_file_tif = r'{0}_fillnodata{1}'.format(base_filename,'.tif')    
    out_file =  r'{0}_fillnodata{1}'.format(base_filename,file_extension)
        
    python_exe = r'C:\Python35\python.exe'
    rootOgrGdal = r'C:\Python35\Lib\site-packages\osgeo'
    gdal_fillnodata = os.path.join(rootOgrGdal, 'scripts','gdal_fillnodata.py')    
    gdal_translate = os.path.join(rootOgrGdal, 'gdal_translate.exe')
    
    command = [python_exe, gdal_fillnodata, '-q', '-md', '1000', in_file, tmp_file_tif]
    cmd(command)

    command = [gdal_translate, '-of', 'AAIGrid', '-a_nodata', '-9999', tmp_file_tif, out_file]
    cmd(command)    
    
    
    return os.path.basename(out_file)
```

```python
asc_files = glob.glob(r'D:\FEWSProjecten\OmsWaddenzee\trunk\fews\Import\hkv\coefficienten\asciiData//*/*/*.asc', recursive=True)
#asc_files = asc_files[0:5]
```

```python
for asc_file in asc_files:
    out_file = func_fillnodata(asc_file)
    print(r'processed {0} into {1}'.format(os.path.basename(asc_file),out_file))
```
