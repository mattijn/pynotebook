```python
# import the cf package
import cf
```

```python
# Read the field from file ncas_data/IPSL-CM5A-LR_r1i1p1_tas_n96_rcp45_mnth.nc
f = cf.read_field('ncas_data/IPSL-CM5A-LR_r1i1p1_tas_n96_rcp45_mnth.nc')
```

```python
# Find the units of the field
# (use the units attribute)
print f.units
```

```python
# Print the field's data array as a numpy array
# (use the array attribute)
f.array
```

```python
# Mask the field where it's values are less than 273.15 Kelvin
# (use the field's where method, with a condition defined with a cf.lt query, and choose missing values cf.masked)
g = f.where(cf.lt(273.15), cf.masked)
```

```python
# Find the number of un-masked points in the original and masked fields
# (use the count method of the field's data attribute)
print f.data.count()
print g.data.count()
```

```python
# Set up inline images and import cf-plot 
%matplotlib inline
import cfplot as cfp
```

```python
# Plot the first time slice of the original and masked fields
# (use cfp.con)
cfp.con(f.subspace[0])
cfp.con(g.subspace[0])
```

```python
# Copy the original field and convert its units to degrees Celcius
c = f.copy()
c.units = 'degC'
```

```python
# Mask te original field where the "celcius" field is less than zero
# (define the condition of where to mask as c<0, where c is the celcius field created above)
h = f.where(c<0, cf.masked)
```

```python
# Check that the two methods of masking give the same result
# (use the field's equals method)
h.equals(g)
```

```python
# Mask the field at all points outside of the tropics
# (use the field's indices method to find the required indices and then use subspace with assignment in index-space
# use the cf.wo function to find locations without (i.e. outside of) a range)
indices= f.indices(latitude=cf.wo(-30, 30))
print indices
g = f.copy()
g.subspace[indices] = cf.masked
```

```python
# Create the same masked field using the where method
# (use the item keyword to the cf.Field.where method to indicate that the condition is applied to 
#  particular coordinates rather than the field's array values)
h = f.where(cf.wo(-30, 30), cf.masked, item='latitude')
```

```python
# Check that the two methods give the same result
h.equals(g)
```

```python
# Mask the original data where it is less than its time average
# (use the < comparison operator between two fields to create the condition describing where
#  the field is less than its mean)
t_mean = f.collapse('T: mean')
g = f.where(f<t_mean, cf.masked)
print g
```

```python
# Plot where the last time slice is less than the temporal average
# (use cfp.con)
cfp.con(g.subspace[-1])
```
