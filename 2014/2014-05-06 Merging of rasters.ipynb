```python
import sys
import subprocess as sp
from osgeo import gdal, ogr
import os
```

```python
def FilesFolder(inGSODFolder, format_end=''):
    """
    All files within one folder
    """
    st_wmo = [os.path.join(root, name)
               for root, dirs, files in os.walk(inGSODFolder)
                 for name in files                 
                 if name.endswith(format_end)]
    return st_wmo

def FilesFolder2(RootFolder, contains='', format_end=''):
    """
    All files within all subfolders
    """
    lists = [os.path.join(root, name)    
                 for root, dirs, files in os.walk(RootFolder)
                   for name in files
                   if contains in name
                     if name.endswith(format_end)]
    return lists
```

```python
folder_in = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\2008\OUTPUT_SummerOnly'
files_psdx = FilesFolder2(folder_in, contains='psdx', format_end='.tif')
files_psdy = FilesFolder2(folder_in, contains='psdy', format_end='.tif')
files_csdxy = FilesFolder2(folder_in, contains='csdxy', format_end='.tif')
files_coh = FilesFolder2(folder_in, contains='coh', format_end='.tif')
files_ph = FilesFolder2(folder_in, contains='ph', format_end='.tif')
```

```python
folder_script= r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\2008\OUTPUT_SummerOnly\script_temp//'
gdal_merge = r'D:\Program Files\QGIS Valmiera\bin//gdal_merge.py'
folder_out = r'D:\Data\0_DAILY_INTERVAL_NDVI_TRMM\InnerMongolia\2008\OUTPUT_SummerOnly\COMB//'

# PSDX
fil_psdx = folder_script+'psdx.txt'
with open(fil_psdx, "w") as thefile:
    for item in files_psdx:
        thefile.write("%s\n" % item)

psdx_out = folder_out+'psdx.tif'

paramsnorm = [gdal_merge, "-o", psdx_out, "-q", "-v", "--optfile", fil_psdx]
norm = sp.Popen(sp.list2cmdline(paramsnorm), shell=True)     
norm.communicate()

# PSDY
fil_psdy = folder_script+'psdy.txt'
with open(fil_psdy, "w") as thefile:
    for item in files_psdy:
        thefile.write("%s\n" % item)
psdy_out = folder_out+'psdy.tif'

paramsnorm = [gdal_merge, "-o", psdy_out, "-q", "-v", "--optfile", fil_psdy]
norm = sp.Popen(sp.list2cmdline(paramsnorm), shell=True)     
norm.communicate("w")

# CSDXY
fil_csdxy = folder_script+'csdxy.txt'
with open(fil_csdxy, "w") as thefile:
    for item in files_csdxy:
        thefile.write("%s\n" % item)
csdxy_out = folder_out+'csdxy.tif'

paramsnorm = [gdal_merge, "-o", csdxy_out, "-q", "-v", "--optfile", fil_csdxy]
norm = sp.Popen(sp.list2cmdline(paramsnorm), shell=True)     
norm.communicate()

# COH
fil_coh = folder_script+'coh.txt'
with open(fil_coh, "w") as thefile:
    for item in files_coh:
        thefile.write("%s\n" % item)
coh_out = folder_out+'coh.tif'

paramsnorm = [gdal_merge, "-o", coh_out, "-q", "-v", "--optfile", fil_coh]
norm = sp.Popen(sp.list2cmdline(paramsnorm), shell=True)     
norm.communicate()        

# PHASE
fil_ph = folder_script+'ph.txt'
with open(fil_ph, "w") as thefile:
    for item in files_ph:
        thefile.write("%s\n" % item)  
ph_out = folder_out+'ph.tif'

paramsnorm = [gdal_merge, "-o", ph_out, "-q", "-v", "--optfile", fil_ph]
norm = sp.Popen(sp.list2cmdline(paramsnorm), shell=True)     
norm.communicate()
```
