```python
from osgeo import gdal
import numpy as np
import matplotlib.pyplot as plt
```

```python
def saveRaster(path, array, dsSource, datatype=3, formatraster="GTiff", nan=None): 
    """
    Datatypes:
    unknown = 0
    byte = 1
    unsigned int16 = 2
    signed int16 = 3
    unsigned int32 = 4
    signed int32 = 5
    float32 = 6
    float64 = 7
    complex int16 = 8
    complex int32 = 9
    complex float32 = 10
    complex float64 = 11
    float32 = 6, 
    signed int = 3
    
    Formatraster:
    GeoTIFF = GTiff
    Erdas = HFA (output = .img)
    OGC web map service = WMS
    png = PNG
    """
    # Set Driver
    format_ = formatraster #save as format
    driver = gdal.GetDriverByName( format_ )
    driver.Register()
    
    # Set Metadata for Raster output
    cols = dsSource.RasterXSize
    rows = dsSource.RasterYSize
    bands = 1#dsSource.RasterCount
    datatype = datatype#band.DataType
    
    # Set Projection for Raster
    outDataset = driver.Create(path, cols, rows, bands, datatype)
    geoTransform = dsSource.GetGeoTransform()
    outDataset.SetGeoTransform(geoTransform)
    proj = dsSource.GetProjection()
    outDataset.SetProjection(proj)
    
    # Write output to band 1 of new Raster and write NaN value
    outBand = outDataset.GetRasterBand(1)
    if nan != None:
        outBand.SetNoDataValue(nan)
    outBand.WriteArray(array) #save input array
    #outBand.WriteArray(dem)
    
    # Close and finalise newly created Raster
    #F_M01 = None
    outBand = None
    proj = None
    geoTransform = None
    outDataset = None
    driver = None
    datatype = None
    bands = None
    rows = None
    cols = None
    driver = None
    array = None
```

```python
depth = np.array([4.5, 4.6, 7.5, 12.3, 20.4, 33.6, 55.4])
path1500 = r'G:\New folder\AWC_CLASS2\sm_retention//1500china_res.tif'
path33 = r'G:\New folder\AWC_CLASS2\sm_retention//33china_res.tif'
ds1500 = gdal.Open(path1500)
ds33 = gdal.Open(path33)
```

```python
bds = ds33.RasterCount
xds = ds33.RasterXSize
yds = ds33.RasterYSize
[bds,xds,yds]
```

```python
for i in range(1,8):
    out = depth[0:i]
    out_arr = np.zeros((out.size,yds,xds))
    for j in range(out.size):
        print(j+1)
        bd33 = ds33.GetRasterBand(j+1)
        ar33 = bd33.ReadAsArray()
        ar_msk_33 = np.ma.masked_where(ar33==-9999.0, ar33)
        
        bd1500 = ds1500.GetRasterBand(j+1)    
        ar1500 = bd1500.ReadAsArray()    
        ar_msk_1500 = np.ma.masked_where(ar1500==-9999.0, ar1500) 
        
        # availbale water content
        awc = ar_msk_33 - ar_msk_1500
        
        # print (awc, out[j-1])
        # out_arr[j-1] = awc*out[j-1]
        out_arr[j] = awc*out[j]
    #msk = ar33[None]
    #ar_msk = np.ma.masked_where(msk==-9999.0,out_arr)
    msk = np.zeros((out.size, ar33.shape[0], ar33.shape[1]))
    for i in range(out.size):
        msk[i-1] = ar33
    ar_msk = np.ma.masked_where(msk == -9999.0, out_arr)    
    ar_msk = np.ma.sum(ar_msk, axis=0)/sum(out)
    #check = out_arr.sum(axis=0)/sum(out)
    print(sum(out))
    path_out = r'C:\Users\Matt\Downloads\AWC_CLASS2\sm_retention//sm_awc_'+str(sum(out)).replace('.','_')+'.tif'
    print(path_out)
    im = plt.imshow(ar_msk, vmin=0.085, vmax=0.185)
    plt.colorbar(im).set_label('AWC cm3*cm-3')
    plt.title('soil depth 0-'+str(sum(out))+' (cm)')
    plt.show()
    print("")
    
    
    #saveRaster(path_out, ar_msk, ds33, datatype=6, nan = 0.0)
```
