```python
import numpy as np
import scipy.signal as signal
import nitime.algorithms as nt_alg
import nitime.utils as nt_ut
import matplotlib.pyplot as pp
```

```python
N = 10000
nfft = np.power(2, int(np.ceil(np.log2(N))) )
NW = 40
W = float(NW)/N
print N, nfft, NW, W
```

```python
s = np.cumsum (np.random.randn(N) )

(b,a) = signal.butter(3, W, btype='lowpass')
slp = signal.lfilter(b,a,s)

pp.figure(figsize=(14,5))
pp.plot(s,'r')
pp.plot(slp, lw=2)
pp.gcf().tight_layout()
pp.grid()
```

```python
s_mod = s * np.cos(2*np.pi*arange(N) * float(200) / N)
slp_mod = slp * np.cos(2*np.pi*np.arange(N) * float(200)/N)
fm = int( np.round(float(200) * nfft / N) )

pp.figure(figsize=(14,5))
pp.plot(s_mod,'r,')
pp.plot(slp_mod, 'b,', lw=2)
pp.gcf().tight_layout()
pp.grid()
```

```python
(dpss, eigs) = nt_alg.dpss_windows(N, NW, 2*NW)
keep = eigs > 0.9
dpss = dpss[keep]; eigs = eigs[keep]
```

```python
# MT method
xk = nt_alg.tapered_spectra(slp_mod, dpss, NFFT=nfft)
mtm_bband1 = np.sum( 2 * (xk[:,fm] * np.sqrt(eigs))[:,None] * dpss, axis=0 )

# Hilbert transform method
hb_bband = signal.hilbert(slp_mod, N=nfft)[:N]

pp.figure(figsize=(14,5))
pp.subplot(211)
pp.plot(slp_mod, 'g', label='signal lowpass modulated')
pp.plot(slp, 'r', lw=5, label='signal lowpass')
pp.plot(np.abs(mtm_bband1), color = 'b', lw=3, label='multitaper baseband estimation')
pp.legend(loc=3,ncol=3,frameon=False)
pp.title('Multitaper Baseband Power')

pp.subplot(212)
pp.plot(slp_mod, 'g', label='signal lowpass modulated')
pp.plot(slp, 'r', lw=5, label='signal lowpass')
pp.plot(np.abs(hb_bband), color='b', lw=3, label='hilbert transform baseband estimation')
pp.legend(loc=3,ncol=3,frameon=False)
pp.title('Hilbert Baseband Power')

pp.gcf().tight_layout()
```

```python
# MT method
xk = nt_alg.tapered_spectra(s_mod, dpss, NFFT=nfft)
w, n = nt_ut.adaptive_weights(xk, eigs, sides='onesided')
mtm_bband2 = np.sum (2 * (xk[:,fm] * np.sqrt(eigs))[:,None] * dpss, axis=0 )

# Hilbert transform method
hb_bband = signal.hilbert(s_mod, N=nfft)[:N]

pp.figure(figsize=(14,5))
pp.subplot(211)
pp.plot(s_mod, 'g', label='signal modulated')
pp.plot(s, 'r', lw=5, label='signal')
pp.plot(np.abs(mtm_bband2), color = 'b', lw=3, label='multitaper baseband estimation')
pp.legend(loc=3,ncol=3,frameon=False)
pp.title('Multitaper Baseband Power')

pp.subplot(212)
pp.plot(s_mod, 'g', label='signal modulated')
pp.plot(s, 'r', lw=5, label='signal')
pp.plot(np.abs(hb_bband), color='b', lw=3, label='hilbert transform baseband estimation')
pp.legend(loc=3,ncol=3,frameon=False)
pp.title('Hilber Baseband Power')
pp.gcf().tight_layout()
```

```python
eigen_coefs = 2 * (xk[:,(fm-100):(fm+100):10] * np.sqrt(w[:,(fm-100):(fm+100):10]))
mtm_fbband = np.sum( eigen_coefs[:,:,None] * dpss[:,None,:], axis=0 )

pp.figure(figsize=(14,5))
pp.plot(s_mod,'g', label='signal modulated')
pp.plot(np.abs(mtm_fbband).T, ls='-', lw=2, label='multitaper freqs estimation')
pp.plot(np.abs(mtm_bband2), color='b', lw=3, label='multitaper baseband estimation')
pp.title('Multitaper Baseband Power: Demodulation Freqs in (fm-100, fm+100)')
pp.legend(loc=3,ncol=4,frameon=False)
pp.gcf().tight_layout()
```
