```python
import pandas as pd
import numpy as np
```

```python
def doy(month, day, year=None):
    """
    Calculation of day of year. If year is provided it will be tested for
    leap years.

    Parameters
    ----------
    month : numpy.ndarray or int32
        Month.
    day : numpy.ndarray or int32
        Day.
    year : numpy.ndarray or int32, optional
        Year.

    Retruns
    -------
    doy : numpy.ndarray or int32
        Day of year.
    """
    daysPast = np.array([0, 31, 60, 91, 121, 152, 182, 213,
                         244, 274, 305, 335, 366])

    day_of_year = daysPast[month - 1] + day

    if year is not None:
        nonleap_years = np.invert(is_leap_year(year))
        day_of_year = day_of_year - nonleap_years + \
            np.logical_and(day_of_year < 60, nonleap_years)

    return day_of_year


def is_leap_year(year):
    """
    Check if year is a leap year.

    Parameters
    ----------
    year : numpy.ndarray or int32

    Returns
    -------
    leap_year : numpy.ndarray or boolean
        True if year is a leap year.
    """
    return np.logical_or(np.logical_and(year % 4 == 0, year % 100 != 0),
                         year % 400 == 0)
```

```python
rng = pd.date_range(start = '2001-01-01',periods = 5113)
ts = pd.TimeSeries(np.random.randn(len(rng)), rng)
ts = ts.cumsum()
ts.plot()
```

```python
doys = doy(ts.index.month, ts.index.day)
```

```python
def anomaly_computation(ts, norm=False):
    if norm == False:
        anom_ts = ts.groupby([ts.index.month, ts.index.day]).apply(lambda g: g - g.mean())
    if norm == True:
        anom_ts = ts.groupby([ts.index.month, ts.index.day]).apply(lambda g: (g - g.mean())/g.std())
    return anom_ts
```

```python
rng = pd.date_range(start = '2001-01-01',periods = 5113)
ts = pd.TimeSeries(np.random.randn(len(rng)), rng)
ts = ts.cumsum()

out = anomaly_computation(ts, norm=True)
out.plot()
```
