```python
import altair as alt
```

```python
url = "https://tsws.hkvservices.nl/mangrove.ws/entry.asmx/Call?function=soilviewer.appfunctions.getdata&parameters={appid:12,name:[%27DKMP_57.5_bit%27,%27DKMP_57.8_bit%27],parameter:[%27Puntdruk%27,%27Lokale%20wrijving%27]}"
```

```python
base = alt.Chart().encode(
    y=alt.Y("depth:Q", scale=alt.Scale(domain=[15,0])), #sort="descending" # this makes the panning also descending
    order=alt.Order("depth:Q"),    
    tooltip=alt.Tooltip([
        "locationname:N", 
        "parametername:N",
        "unit:N",
        "value:Q", 
        "depth:Q"
    ])
)

line = base.mark_line().encode(
    x=alt.X("value:Q"),
    color=alt.Color("parametername:N")
).properties(
    width=150, 
    height=150
).interactive(bind_x=False, )

nearest = alt.selection_single(on="mouseover", nearest=True, empty='none')
rule = base.mark_rule().encode(
    color=alt.condition(nearest, alt.value('black'), alt.value('transparent'))
).add_selection(
    nearest
)

chart = alt.layer(line, rule, data=url)
chart.facet(
    row=alt.Row("parametername:N", title=None), 
    column="locationname:N"
).resolve_scale(
    x="independent"
)
```
