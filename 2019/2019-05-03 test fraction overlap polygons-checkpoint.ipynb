```python
import geopandas as gpd
from collections import OrderedDict
import pandas as pd
import itertools
from shapely.ops import unary_union
from shapely import geometry
from IPython.display import SVG
%matplotlib inline
```

```python
def dissolve_agg_column(gdf, by_column='ONDPEILGEB', into_column='CODE'):
    children = gdf[~(gdf[into_column] == gdf[by_column])]
    
    parents = gpd.GeoDataFrame()
    for pg_id in children[by_column].unique():
        if pg_id != None:
            children_sel = children[children[by_column] == pg_id]
            parent = gdf[(gdf[into_column] == pg_id) & (gdf[by_column] == pg_id)]

            if parent.shape[0] != 1:    
                print('error in {} {}'.format(pg_id, children_sel[[into_column,by_column]]))

            else:
                # do unary_union of children in parent
                geom_tmp = unary_union(children_sel.append(parent)['geometry'])
                parent['geometry'].iloc[0] = geom_tmp            
                parents = parents.append(parent)

        else:
            orphans = children[children[by_column].isnull()]

    parents_childless = gdf[gdf[into_column] == gdf[by_column]]
    childless = parents_childless.drop(parents.index)

    gdf_dissolved = childless.append(parents)
    return gdf_dissolved
```

```python
gdf_base_dissolved = dissolve_agg_column(gdf_base, by_column='ONDPEILGEB', into_column='CODE')
#gdf_update_dissolved = dissolve_agg_column(gdf_update, by_column='ONDPEILGEB', into_column='PGIDENT')
```

```python
geoms_base = []
for idx, row in gdf_base_dissolved.iterrows():
    obj = {'file':'base',
           'geom':row['geometry'],
           'id': row['CODE']
          }
    geoms_base.append(obj)
    
geoms_update = []
for idx, row in gdf_update.iterrows():
    obj = {'file':'update',
           'geom':row['geometry'],
           'id': row['GPGIDENT']
          }
    geoms_update.append(obj)    
```

```python
geom_pairs = list(itertools.product(geoms_base, geoms_update))
```

```python
check_obj = dict.fromkeys(['ID_BASE', 'ID_UPDATE', 'ID_EQUAL', 'OVERLAP', 'FRACTION_OVERLAP'])
checks = []

for pair in geom_pairs:  
    obj_base = pair[0]
    obj_update = pair[1]
    
    check_obj = OrderedDict.fromkeys(['ID_BASE', 'ID_UPDATE', 'ID_EQUAL', 
                                      'OVERLAP', 'FRACTION_OVERLAP_BASE', 'FRACTION_OVERLAP_UPDATE'])   
    check_obj.update({'ID_BASE': obj_base['id']})
    check_obj.update({'ID_UPDATE': obj_update['id']})
    check_obj.update({'ID_EQUAL': obj_base['id'] == obj_update['id']})
    check_obj.update({'OVERLAP': obj_base['geom'].overlaps(obj_update['geom'])})
        
    if check_obj['OVERLAP']:      
        geom_int = obj_base['geom'].intersection(obj_update['geom'])
        opp_int = geom_int.area
        opp_base = obj_base['geom'].area
        opp_update = obj_update['geom'].area
        perc_base = opp_int / opp_base
        perc_update = opp_int / opp_update        
               
        check_obj.update({'FRACTION_OVERLAP_BASE': perc_base})
        check_obj.update({'FRACTION_OVERLAP_UPDATE': perc_update})
    
    checks.append(check_obj)
```

```python
df = pd.DataFrame(checks)
```

```python
overlaps = df[df['OVERLAP'] == True]
overlaps[overlaps['FRACTION_OVERLAP_BASE'] < 10].head(10)
```

```python
def compare_id(row_idx, gdf_base, gdf_update, base_col='CODE', update_col='GPGIDENT'):
    row = overlaps.loc[row_idx]
    
    g1 = gdf_base[gdf_base[base_col]==row['ID_BASE']].iloc[0]['geometry']
    g2 = gdf_update[gdf_update[update_col]==row['ID_UPDATE']].iloc[0]['geometry']
    
    svg_string = geometry.GeometryCollection([g1, g2])._repr_svg_()
    svg_string = svg_string.replace('stroke="#555555"', 'stroke="red"', 1)
    svg_string = svg_string.replace('fill="#ff3333"', 'fill="red"', 1)


    return g1, g2, SVG(svg_string)    
```
