```python
import pandas as pd
import geopandas as gpd
import altair as alt
```

```python
gdf = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
eea_countries = list(['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Latvia', 'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Norway', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'United Kingdom'])
gdf_names = gdf.name.tolist()
matching_names = list(set(eea_countries).intersection(gdf_names))
gdf_eea = gdf[gdf.name.isin(matching_names)]
gdf_eea.reset_index(drop=True, inplace=True)
```

```python
alt.Chart(gdf_eea).mark_geoshape().encode(tooltip=['name'])
```

```python
gdf_eea.to_file('europe_eea.shp')
```
