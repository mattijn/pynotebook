```python
import numpy as np
import pandas as pd
import altair as alt

x, y = np.meshgrid(np.arange(30), np.arange(30))
z = np.exp(-(x - 14)**2 / (5 ** 2) - (y - 22) ** 2 / (5 ** 2))
data = pd.DataFrame({
    'x': x.ravel(),
    'y': y.ravel(),
    'z': z.ravel()
})

alt.Chart(data).mark_rect().encode(
  x='x:O',
  y='y:O',
  fill='z:Q'
).configure_scale(bandPaddingInner=0)
```
