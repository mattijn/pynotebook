```python
from io import StringIO
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import topojson
import hkvportal
import io
import altair as alt
import pandas as pd
dp = hkvportal.Service('https://tsws.hkvservices.nl/mangrove.ws/', '236ebLIeCDdnKbfGe956IvZetIv1')
```

```python
geom = gpd.read_file('europe_eea.shp')
```

```python
gdf = gpd.read_file('NO2_EU_EEA_geojson_collection.geo.json', driver='GeoJSON')
```

```python
gdf.head()
```

```python
data = pd.merge(gdf, geom[['geometry', 'name']], how='inner', on='name', suffixes=('_x',''))
```

```python
from shapely.geometry.base import BaseMultipartGeometry
from shapely.geometry.polygon import orient as orient_
from shapely.geometry import Polygon

def orient(geom, sign=1.0):
    if isinstance(geom, BaseMultipartGeometry):
        return geom.__class__(
            list(
                map(
                    lambda geom: orient(geom, sign),
                    geom.geoms,
                )
            )
        )
    if isinstance(geom, (Polygon,)):
        return orient_(geom, sign)
    return geom
```

```python
#from shapely.ops import orient # version >=1.7a2
data.geometry = data.geometry.apply(orient, args=(-1,))
```

```python
data.head()
```

```python
dates = data.date.unique()
dates_dropdown = alt.binding_select(options=dates.tolist())
dates_select = alt.selection_single(empty='none', fields=['date'], bind=dates_dropdown, name="Dates", init={'date':dates[-1]})#, on="click[event.shiftKey&!event.shiftKey]", clear=False)
country_select = alt.selection_single(empty='none', fields=['name'], init={'name':'Netherlands'}, on='mouseover')
country_date_highlight = alt.selection_single(empty='none', fields=['name', 'date'], on='mouseover')
```

```python
width = 300
height = 300

mapplt = alt.Chart().mark_geoshape(stroke='black').encode(
    color=alt.Color('median:Q', scale=alt.Scale(type='quantize', scheme=alt.SchemeParams('blueorange', count=5), domain=[0,0.00016]),legend=alt.Legend(title='NO₂ mol/m²', format='.1s')),
    tooltip=['name',alt.Tooltip('median', format='.4s')],
    strokeWidth=alt.condition(country_select, alt.StrokeWidthValue(2), alt.StrokeWidthValue(1))
).properties(
    width=width,
    height=height,
    projection={'type': 'conicConformal', 'rotate': [-20, 0.0], 'center':[-11, 54.0],'scale':400, 'precision':0.1, 'translate':[width/2,height/2]}
).add_selection(
    dates_select
).transform_filter(
    dates_select
).add_selection(country_select)
```

```python
baseplt = alt.Chart().mark_geoshape(stroke='black', fill='white',strokeWidth=1).encode(
).properties(
    width=width,
    height=height,
    projection={'type': 'conicConformal', 'rotate': [-20, 0.0], 'center':[-11, 54.0],'scale':400, 'precision':0.1, 'translate':[width/2,height/2]}
)
```

```python
titleplt = alt.Chart().mark_text(size=12).encode(
    text='date:N'
).add_selection(
    dates_select
).transform_filter(
    dates_select
)
```

```python
boxplt = alt.Chart().mark_boxplot().encode(
    y=alt.Y('date:O', title=None),
    x=alt.X('median:Q', title='NO₂ mol/m²', axis=alt.Axis(format='.1s')),
    color=alt.condition(dates_select, alt.value('#FF9C33'), alt.value('#77B2D4')),
    tooltip=['name',alt.Tooltip('median', format='.4s')]
).properties(width=width
).add_selection(
    dates_select
)
```

```python
circleplt = alt.Chart().mark_circle(stroke='black'
).encode(    
    x=alt.X('date:O', axis=alt.Axis(labelAngle=0), title=None),
    y=alt.Y('name:N', title=None),
    size=alt.Size('median:Q',
        scale=alt.Scale(range=[0, 500], type='symlog'),
        legend=alt.Legend(title='NO₂ mol/m²', format='.1s')
    ),
    fill=alt.Fill('median:Q', scale=alt.Scale(type='quantize', scheme=alt.SchemeParams('blueorange', count=5), domain=[0,0.00016]), legend=None),
    tooltip=['date:O','name:N',alt.Tooltip('median', format='.4s')],
    strokeWidth=alt.condition(country_select, alt.StrokeWidthValue(2), alt.StrokeWidthValue(1)),
    fillOpacity=alt.condition(country_date_highlight, alt.FillOpacityValue(0.6), alt.FillOpacityValue(1))
).add_selection(
    dates_select
).add_selection(
    country_date_highlight
)
```

```python
chart = alt.hconcat(alt.vconcat(titleplt,alt.layer(baseplt,mapplt), boxplt),circleplt, data=data)
```

```python
# export visualization as an 'in-memory text stream'
export_html = io.StringIO()
chart.save(export_html, format='html')
```
