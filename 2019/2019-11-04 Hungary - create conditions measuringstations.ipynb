```python
import pandas as pd
import geopandas as gpd
from glob import glob
%matplotlib inline
```

```python
station1 = r"D:\Projects\Pr\2505.12\Eduvizig\Sensor stations (waterstandlocaties)\stations_bak.shp"
station2 = r"D:\Projects\Pr\2505.12\Eduvizig\Sensor stations (waterstandlocaties)\stations.shp"
gdf_station1 = gpd.read_file(station1)
gdf_station2 = gpd.read_file(station2)
```

```python
gdf_stations = pd.merge(gdf_station1, gdf_station2, left_on='name', right_on='name')
gdf_stations.head()
```

```python
df_all = pd.DataFrame()
collection = []
for row in gdf_stations.iterrows():
    loc = row[1]
    loc_info = {
        'stationid':loc.id_y,
        'min':int(loc['min_waterl']), 
        'max':int(loc['max_waterl']), 
        '0_thd': loc['threshold_'], 
        '1_thd': loc['threshol_1'], 
        '2_thd': loc['threshol_2']
    }
    thd_level = {
        '0_thd': 0/3, 
        '1_thd': 1/3, 
        '2_thd': 2/3
    }
    
    if loc_info['0_thd'] == 0 and loc_info['1_thd'] == 0 and loc_info['2_thd'] == 0:
        waterlevel_divid = (loc_info['max'] - loc_info['min']) / 8
        loc_info['0_thd'] = loc_info['min'] + (5 * waterlevel_divid)
        loc_info['1_thd'] = loc_info['min'] + (6 * waterlevel_divid)        
        loc_info['2_thd'] = loc_info['min'] + (7 * waterlevel_divid)                
    
    print(loc['name'], loc.id_y, loc_info)
    

    for key in loc_info.keys():
        dic = {}
        dic.update({'objectid': loc_info['stationid']})
        dic.update({'objecttype': 'measuringstation'})

        if key == '0_thd':
            dic.update({'stateid': 1})                
            dic.update({'upperboundary':loc_info[key]})
            dic.update({'name': '5XP1Y'})
            dic.update({'description': '5 times a year'})
            dic.update({'color': '#daeef3'})     
            collection.append(dic)
        if key == '1_thd':
            dic.update({'stateid': 2})                        
            dic.update({'upperboundary':loc_info[key]})
            dic.update({'name': '5XP1J'})
            dic.update({'name': '1XP30Y'})
            dic.update({'description': '1 times a 30 year'})        
            dic.update({'color': '#4bacc6'})        
            collection.append(dic)
        if key == '2_thd':
            dic.update({'stateid': 3})
            dic.update({'upperboundary':loc_info[key]})
            dic.update({'name': '1XP100Y'})
            dic.update({'description': '1 times a 100 year'})        
            dic.update({'color': '#31849b'})          
            collection.append(dic)
        if key == 'max':
            dic.update({'stateid': 4})        
            dic.update({'upperboundary':loc_info[key]})
            dic.update({'name': '1XP1000Y'})
            dic.update({'description': '1 times a 1000 year'})        
            dic.update({'color': '#002060'})          
            collection.append(dic)    
```

```python
df = pd.DataFrame(collection)
```

```python
df = df[['stateid', 'objectid', 'objecttype', 'upperboundary', 'name', 'description', 'color']].sort_values(['objectid','stateid'])
df.head()
```
