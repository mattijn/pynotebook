```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
path_in = r"D:\Projects\Pr\3915.10\Data_Viewer\df_volume_sand_update20190603.csv"
df = pd.read_csv(path_in, sep=';', index_col='Index_Total')#'Index_Total', )
```

```python
df.columns[0]
```

```python
df.head()
```

```python
geometry = [Point(xy) for xy in zip(df.X, df.Y)]
```

```python
df = df.drop(['Unnamed: 0','X', 'Y', 'Jarkus', 'Angle_Deg','Angle_Rad'], axis=1)
crs = {'init': 'epsg:28992'}
gdf = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
gdf = gdf.fillna(0)
```

```python
gdf.to_crs(epsg=4326, inplace=True)
```

```python
gdf.crs
```

```python
%matplotlib inline
gdf.plot()
```

```python
gdf.head()
```

```python
cols = gdf.columns
```

```python
cols[0]
```

```python
new_cols = []
for col in cols:
    col = col.replace('Volume>','')
    col2 = col.replace('mzeespiegelstijging','')
    try:
        #float(col2)
        col2 = "{0:.2f}".format(float(col2))
        col2 = float(col2)
        #col2=col2.replace('.', '_')
        new_cols.append(col2)
    except:
        new_cols.append(col)    
```

```python
gdf.columns = new_cols
```

```python
gdf.head()
```

```python
gdf_melt = gdf.melt(id_vars=['geometry'])
```

```python
gdf_melt.head()
```

```python
gdf_melt.variable = gdf_melt.variable.astype(float)
```

```python
gdf_melt.head()
```

```python
gdf_melt.variable = (gdf_melt.variable * 100).astype(int)
```
